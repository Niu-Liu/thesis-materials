I describle the development of Ka-band celestial frame.

```python
from astropy.table import Table, Column
from astropy.time import Time

from matplotlib import pyplot as plt
from matplotlib.ticker import MultipleLocator
%matplotlib inline
```

```python
ka_cat = Table.read("../xka-band-crf/statistics-offset.txt", format="ascii")
ka_cat
```

```python
epoch = Time(ka_cat["epoch"], scale="utc")
epoch = Column(epoch.jyear)
ka_cat["epoch"] = epoch
```

```python
fig, ax = plt.subplots()

ax.plot(ka_cat["epoch"], ka_cat["num_sess"], "b-s", lw=0.5)

ax.axis([2006.5, 2020, 0, 200])

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.set_ylabel("No. Sessions", fontsize=12)
ax.set_xlabel("Year", fontsize=12)
ax.xaxis.set_minor_locator(MultipleLocator(1))

plt.tight_layout()
plt.savefig("ka-band-sess.eps", hbox="tight")
```

```python
fig, ax = plt.subplots()

ax.plot(ka_cat["epoch"], ka_cat["num_sou"], "b-s", lw=0.5, label="All")
ax.plot(ka_cat["epoch"], ka_cat["num_com_sou"], "r-^", lw=0.5, label="Common to S/X")

ax.axis([2006.5, 2020, 200, 700])

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.set_ylabel("No. Sources", fontsize=12)
ax.set_xlabel("Year", fontsize=12)
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.legend()

plt.tight_layout()
plt.savefig("ka-band-sou.eps", hbox="tight")
```

```python
fig, ax = plt.subplots()

ax.plot(ka_cat["epoch"], ka_cat["wrms_ra"], "b-s", lw=0.5, label="$\Delta\\alpha\cos\delta$")
ax.plot(ka_cat["epoch"], ka_cat["wrms_dec"], "r-^", lw=0.5, label="$\Delta\delta$")

ax.axis([2006.5, 2020, 120, 450])

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.set_ylabel("WRMS (${\\rm \mu as}$)", fontsize=12)
ax.set_xlabel("Year", fontsize=12)
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.legend()

plt.tight_layout()
plt.savefig("ka-band-wrms.eps", hbox="tight")
```
