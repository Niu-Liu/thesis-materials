```python
from astropy import units as u
from astropy.table import Table, Column
from astropy.time import Time
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord
import numpy as np
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
from my_progs.catalog.read_icrf import read_icrf1, read_icrf2, read_icrf3
```

```python
i1 = read_icrf1()

i1d = i1[i1["type"] == "D"]
i1c = i1[i1["type"] == "C"]
i1o = i1[i1["type"] == "O"]
```

```python
i2 = read_icrf2()

i2d = i2[i2["type"] == "D"]
i2v = i2[i2["type"] == "V"]
i2n = i2[i2["type"] == "N"]
```

```python
i3x = read_icrf3(wv="sx")

i3xd = i3x[i3x["type"] == "D"]
i3xo = i3x[i3x["type"] != "D"]
```

```python
i3k = read_icrf3(wv="k")

i3kd = i3k[i3k["type"] == "D"]
i3ko = i3k[i3k["type"] != "D"]
```

```python
i3ka = read_icrf3(wv="xka")

i3kad = i3ka[i3ka["type"] == "D"]
i3kao = i3ka[i3ka["type"] != "D"]
```

```python
# Distribution of formal errors of VLBI position.
ra1 = coord.Angle(i1d["ra"]).wrap_at(180*u.degree)
dec1 = coord.Angle(i1d["dec"])

ra2 = coord.Angle(i1c["ra"]).wrap_at(180*u.degree)
dec2 = coord.Angle(i1c["dec"])

ra3 = coord.Angle(i1o["ra"]).wrap_at(180*u.degree)
dec3 = coord.Angle(i1o["dec"])


fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111, projection="mollweide")


ax.scatter(ra1.radian, dec1.radian, marker="^", 
            s=5, facecolors="None", edgecolors="red", lw=0.3, label="Defining")  

ax.scatter(ra2.radian, dec2.radian, marker="o", 
            s=5, facecolors="None", edgecolors="b", lw=0.3, label="Candidate")  

ax.scatter(ra3.radian, dec3.radian, marker="s", 
            s=5, facecolors="None", edgecolors="grey", lw=0.3, label="Other")  


ax.set_xticklabels(["14h", "16h", "18h", "20h", "22h", "0h", "2h", "4h", "6h", "8h", "10h"])

ax.grid(True)
ax.legend(fontsize="x-small")

plt.tight_layout()
plt.savefig("icrf1.eps")
```
