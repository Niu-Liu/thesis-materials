In this notebook I will remove the global difference amongst catalogs first and then re-look at the positional offset.

The global difference is modeled as a 16-parameter transformation.



Again, get the quasar positions from the origin catalogs.

```python
from astropy.table import Table, join, Column
from astropy import units as u
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, NullFormatter
%matplotlib inline
```

```python
comsou = Table.read("../logs/common-soource-position.fits")
comsou
```

Then I removed all the global systematics using the 16-parameter transformation.

The transformation parameters were estimated in the previous work.

```python
# Read the transformation parameters
vsh02_k = Table.read("../logs/icrf3_sx_k_vsh02_all.log", format="ascii")
par_k = np.array(vsh02_k["Estimate"]) / 1000  # uas -> mas

vsh02_xka = Table.read("../logs/icrf3_sx_xka_vsh02_all.log", format="ascii")
par_xka = np.array(vsh02_xka["Estimate"]) / 1000  # uas -> mas

vsh02_opt = Table.read("../logs/icrf3sx_gaiadr2_vsh02_all.log", format="ascii")
# The origin offset was calculated in the sense of ICRF3 S/X - Gaia DR2
par_opt = -np.array(vsh02_opt["Estimate"]) / 1000  # uas -> mas
```

```python
from my_progs.catalog.vsh_deg2_cor import vsh_deg02_fitting, residual_calc02
from my_progs.catalog.pos_diff import nor_sep_calc


# convert unit from degree to rad
RA = np.array(comsou["ra"].to(u.radian))
DE = np.array(comsou["dec"].to(u.radian))

# convert Astropy.table.Column to the np.array
# K-band
dra_k = np.array(comsou["dra_k"])
ddec_k = np.array(comsou["ddec_k"])
dra_err_k = np.array(comsou["dra_err_k"])
ddec_err_k = np.array(comsou["ddec_err_k"])
dra_ddec_cov_k = np.array(comsou["dra_ddec_cov_k"])

# X/Ka-band
dra_xka = np.array(comsou["dra_xka"])
ddec_xka = np.array(comsou["ddec_xka"])
dra_err_xka = np.array(comsou["dra_err_xka"])
ddec_err_xka = np.array(comsou["ddec_err_xka"])
dra_ddec_cov_xka = np.array(comsou["dra_ddec_cov_xka"])

# Optical (Gaia) band
dra_opt = np.array(comsou["dra_opt"])
ddec_opt = np.array(comsou["ddec_opt"])
dra_err_opt = np.array(comsou["dra_err_opt"])
ddec_err_opt = np.array(comsou["ddec_err_opt"])
dra_ddec_cov_opt = np.array(comsou["dra_ddec_cov_opt"])

# Re-calculate the positional offset
dra_kn, ddec_kn = residual_calc02(dra_k, ddec_k, RA, DE, par_k)
dra_xkan, ddec_xkan = residual_calc02(dra_xka, ddec_xka, RA, DE, par_xka)
dra_optn, ddec_optn = residual_calc02(dra_opt, ddec_opt, RA, DE, par_opt)

# Now re-calculate the normalized difference
ang_sep_kn, Xa_kn, Xd_kn, X_kn = nor_sep_calc(
    dra_kn, dra_err_k, ddec_kn, ddec_err_k, dra_ddec_cov_k/dra_err_k/ddec_err_k)
ang_sep_xkan, Xa_xkan, Xd_xkan, X_xkan = nor_sep_calc(
    dra_xkan, dra_err_xka, ddec_xkan, ddec_err_xka,
    dra_ddec_cov_xka/dra_err_xka/ddec_err_xka)
ang_sep_optn, Xa_optn, Xd_optn, X_optn = nor_sep_calc(
    dra_optn, dra_err_opt, ddec_optn, ddec_err_opt,
    dra_ddec_cov_opt/dra_err_opt/ddec_err_opt)
```

```python
# np.array -> astropy.table.Column
dra_kn = Column(dra_kn, name="dra_kn")
ddec_kn = Column(ddec_kn, name="ddec_kn")
ang_sep_kn = Column(ang_sep_kn, name="ang_sep_kn")
Xa_kn = Column(Xa_kn, name="nor_ra_kn")
Xd_kn = Column(Xd_kn, name="nor_dec_kn")
X_kn = Column(X_kn, name="nor_sep_kn")
dra_xkan = Column(dra_xkan, name="dra_xkan")
ddec_xkan = Column(ddec_xkan, name="ddec_xkan")
ang_sep_xkan = Column(ang_sep_xkan, name="ang_sep_xkan")
Xa_xkan = Column(Xa_xkan, name="nor_ra_xkan")
Xd_xkan = Column(Xd_xkan, name="nor_dec_xkan")
X_xkan = Column(X_xkan, name="nor_sep_xkan")
dra_optn = Column(dra_optn, name="dra_optn")
ddec_optn = Column(ddec_optn, name="ddec_optn")
ang_sep_optn = Column(ang_sep_optn, name="ang_sep_optn")
Xa_optn = Column(Xa_optn, name="nor_ra_optn")
Xd_optn = Column(Xd_optn, name="nor_dec_optn")
X_optn = Column(X_optn, name="nor_sep_optn")

# Add new columns
comsou.add_columns([dra_kn, ddec_kn, ang_sep_kn, Xa_kn, Xd_kn, X_kn,
                     dra_xkan, ddec_xkan, ang_sep_xkan, Xa_xkan, Xd_xkan, X_xkan,
                     dra_optn, ddec_optn, ang_sep_optn, Xa_optn, Xd_optn, X_optn])
```

```python
def pa_calc(dra, ddec):
    """Calculate positional angle from positional offset.

    Parametes
    ---------
    dra : ndarray
        positional difference in R.A.(times cos(decl.))
    ddec : ndarray
        positional difference in declination

    Returns
    -------
    A : ndarray
        Angle (in degree) of positional offset vector towards to y-axis clockwisely
    """

    # Direction of positional offset vector from the y-axis count-clockwisely
    # in degree
    A = np.rad2deg(np.arctan2(dra, ddec))
    
    A = np.where(A>0, A, A+360)

    return A
```

```python
# Calculate the positional angle of the positional offset
PA_k = pa_calc(comsou["dra_kn"], comsou["ddec_kn"])
PA_xka = pa_calc(comsou["dra_xkan"], comsou["ddec_xkan"])
PA_opt = pa_calc(comsou["dra_optn"], comsou["ddec_optn"])
```

```python
# Histogram of angular separation
fig, ax = plt.subplots(figsize=(8, 6))

bins_set = np.linspace(0, 360, 25)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
ax.hist([PA_k, PA_xka, PA_opt],
        bins_set, histtype="step", fill=False,
        color=["r", "b", "g"],
        label=["K $-$ S/X", "X/Ka $-$ S/X", "Gaia $-$ S/X"],
        weights=[wgt, wgt, wgt])

ax.set_xticks(np.arange(0, 361, 30))
ax.set_ylabel("%", fontsize=12)
ax.set_xlabel("Position Angle [$\circ$]", fontsize=12)

ax.legend(loc="upper left")
# ax.grid()
ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")
```

```python
def inclu_angle_calc(dra1, ddec1, dra2, ddec2):
    """Calculate included angle between two positional offset vectors.


    Parametes
    ---------
    dra1, ddec1: ndarray
        components of the fisrt positional offset vector
    dra2, ddec2: ndarray
        components of the second positional offset vectors

    Returns
    -------
    ang : ndarray
        Angle (in degree) between positional offset vectors
    """

    cosine = dra1 * dra2 + ddec1 * ddec2
    mod1 = np.sqrt(dra1**2 + ddec1**2)
    mod2 = np.sqrt(dra2**2 + ddec2**2)

    ang = np.rad2deg(np.arccos(cosine/mod1/mod2))

    return ang
```

```python
# Calculate the angle between positional offset vectors
ang_k_opt = inclu_angle_calc(
    comsou["dra_kn"], comsou["ddec_kn"], comsou["dra_optn"], comsou["ddec_optn"])
ang_xka_opt = inclu_angle_calc(
    comsou["dra_xkan"], comsou["ddec_xkan"], comsou["dra_optn"], comsou["ddec_optn"])
ang_xka_k = inclu_angle_calc(
    comsou["dra_kn"], comsou["ddec_kn"], comsou["dra_xkan"], comsou["ddec_xkan"])
```

```python
# Histogram of angular separation
fig, ax = plt.subplots(figsize=(8, 6))

bins_set = np.linspace(0, 180, 25)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
ax.hist([ang_k_opt, ang_xka_opt, ang_xka_k],
        bins_set, histtype="step", fill=False,
        color=["r", "b", "g"],
        label=["between K$-$S/X and Gaia$-$S/X", 
               "between X/Ka$-$S/X and Gaia$-$S/X", 
               "between X/Ka$-$S/X and K$-$S/X"],
        weights=[wgt, wgt, wgt])

ax.set_xticks(np.arange(0, 181, 30))
ax.set_ylabel("%", fontsize=12)
ax.set_xlabel("Angle between position offset vectors [$^\circ$]", fontsize=12)

ax.legend()
# ax.grid()
ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")
```
