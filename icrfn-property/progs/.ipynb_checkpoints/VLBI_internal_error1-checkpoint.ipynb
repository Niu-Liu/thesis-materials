```python
# Load data
datfile = ("../data/odd-even.cat_diff")
soulist = np.genfromtxt(datfile, usecols=(0,), dtype=str)
RA, Dec, dRAc, dDec, dRAc_err, dDec_err = np.genfromtxt(
    datfile, usecols=range(1, 7), unpack=True)
num_ses_mean, num_obs_mean = np.genfromtxt(
    datfile, dtype=int, usecols=range(7, 9), unpack=True)
```

```python
# Find VCS sources

# ICRF2 data
_, ivsn, _, _, _, _, _, _, _, Flag = read_icrf2("../data/icrf2.dat")

# VCS source
con = (Flag == "V")
icrf2_vcs = ivsn[con]
    
# Find defining sources
vcssou = [sou for sou in soulist if sou in icrf2_vcs]
        
vcssou = np.asarray(vcssou)

print("There are %d common sources, %d of which are VCS sources." % (soulist.size, vcssou.size))
```

```python
# Plot for the difference
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

bins = np.arange(-5, 5, 0.2)

ax0.hist(dRAc, bins, histtype='bar', rwidth=0.8)
ax0.set_xlabel("mas")
ax0.set_title('$\Delta \\alpha^*$')

ax1.hist(dDec, bins, histtype='bar', rwidth=0.8)
ax1.set_xlabel("mas")
ax1.set_title('$\Delta \delta$')

fig.tight_layout()
```

```python
# scale the position offset by their uncertainty
nor_dRAc = dRAc / dRAc_err
nor_dDec = dDec / dDec_err

# Plot for the normalized difference
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

bins = np.arange(-10., 10., 0.5)

ax0.hist(nor_dRAc, bins, histtype='bar', rwidth=0.8)
ax0.set_xlabel("$\\Delta\\alpha^*/\sigma_{\\Delta\\alpha^*}$")

ax1.hist(nor_dDec, bins, histtype='bar', rwidth=0.8)
# ax1.set_xlabel("mas")
ax1.set_title('$X_{\delta}$')

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
minorLocator = MultipleLocator(1)
ax0.xaxis.set_minor_locator(minorLocator)
ax1.xaxis.set_minor_locator(minorLocator)

ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')

fig.tight_layout()
```

```python
# VCS source
con = (Flag == "V")
icrf2_vcs = ivsn[con]
    
# Find defining sources
sou_flg = []
for sou in soulist:
    if sou in icrf2_vcs:
#         sou_flg.append(False)
        sou_flg.append(True)
    else:
#         sou_flg.append(True)
        sou_flg.append(False)
            
sou_flg = np.asarray(sou_flg)

# nor_dRAc_novcs = nor_dRAc[sou_flg]
# nor_dDec_novcs = nor_dDec[sou_flg]
nor_dRAc_vcs = nor_dRAc[sou_flg]
nor_dDec_vcs = nor_dDec[sou_flg]

# print("Non-VCS Source number: ", nor_dRAc_novcs.size)
print("VCS Source number: ", nor_dRAc_vcs.size)
```

```python
# Plot for the difference
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

bins = np.arange(-5, 5, 0.2)

ax0.hist(dRAc[sou_flg], bins, histtype='bar', rwidth=0.8)
ax0.set_xlabel("mas")
ax0.set_title('$\Delta \\alpha^* (Non-VCS)$')

ax1.hist(dDec[sou_flg], bins, histtype='bar', rwidth=0.8)
ax1.set_xlabel("mas")
ax1.set_title('$\Delta \delta$ (Non-VCS)')

fig.tight_layout()
```

```python
# scale the position offset by their uncertainty
nor_dRAc = dRAc / dRAc_err
nor_dDec = dDec / dDec_err

# Plot for the normalized difference
# fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))
fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(4, 6))

bins = np.arange(-10., 10., 0.5)

ax0.hist(nor_dRAc, bins, histtype='bar', rwidth=0.8, label="All")
ax0.hist(nor_dRAc_vcs, bins, histtype='bar', rwidth=0.8, label="VCS")
ax0.set_xlabel("$\\Delta\\alpha\cos\delta/\sigma_{\\Delta\\alpha\cos\delta}$")
ax0.set_ylabel("Number of Sources")
ax0.set_ylim([0, 450])
ax0.legend()

ax1.hist(nor_dDec, bins, histtype='bar', rwidth=0.8, label="All")
ax1.hist(nor_dDec_vcs, bins, histtype='bar', rwidth=0.8, label="VCS")

# ax1.set_xlabel("mas")
# ax1.set_title('$X_{\delta}$')
ax1.set_ylim([0, 450])
ax1.set_xlabel("$\Delta\delta/\sigma_{\Delta\delta}$")
ax1.set_ylabel("Number of Sources")
# ax1.legend()

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
minorLocator = MultipleLocator(1)
ax0.xaxis.set_minor_locator(minorLocator)
ax1.xaxis.set_minor_locator(minorLocator)

minorLocator = MultipleLocator(10)
ax0.yaxis.set_minor_locator(minorLocator)
ax1.yaxis.set_minor_locator(minorLocator)

ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')


fig.tight_layout()
# plt.subplots_adjust(wspace=0.15)
plt.savefig("../plots/decimation-offset-dist.eps")
```

```python
# Calculate the offset and normalized seperation
rho = sqrt(dRAc**2 + dDec**2)
nor_rho = sqrt(nor_dRAc**2 + nor_dDec**2)
rho_err = sqrt(dRAc_err**2 + dDec_err**2)

fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

bins = np.arange(0., 10., 0.5)

ax0.hist(rho, bins, histtype='bar', rwidth=0.8)
ax0.set_title('$\\rho$ (mas)')
# ax0.set_ylabel("Count")

# ax1.hist(nor_rho, bins, histtype='bar', rwidth=0.8)
ax1.hist(rho / rho_err, bins, histtype='bar', rwidth=0.8)
ax1.set_title('$X$')

fig.tight_layout()
```

```python
# Cumulative plot
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

bins = np.arange(0., 10., 0.1)

ax0.hist(rho, bins, histtype='step', density=True, cumulative=True, rwidth=0.8)
ax0.set_title('$\\rho$ (mas)')
ax0.set_ylim([0, 1.])

# ax1.hist(nor_rho, bins, histtype='step', density=True, cumulative=True, rwidth=0.8)
ax1.hist(rho / rho_err, bins, histtype='step', density=True, cumulative=True, rwidth=0.8)
ax1.set_title('$X$')
ax1.set_ylim([0, 1.])

fig.tight_layout()
```
