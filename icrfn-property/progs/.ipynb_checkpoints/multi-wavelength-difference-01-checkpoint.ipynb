Compare the multi-wavelength positions of quasars
-------
by Niu

With the ICRF3 and Gaia DR2 catalog now we have the quasar position measured indepeddently at four different bands:
S/X, K, and X/Ka from VLBI and optical one from Gaia DR2.

Comparing these four positions may be interesting for studying the wavalength-dependency of quasar position and astrophysical properties of AGNs.

Here I use the S/X position as the reference position (assume this position is the most accurate) and compute the position offsets of other three bands.

Several perspects are investigated in this note, as listed below.

- dependency of positional offset on the R.A. and decl.;
- statistics of angular separation (denoted as $\rho$) and normalized separation (denoted as $X$);
- statistical correlation amongst position angles (PAs) of positional offset vector;
- statistical correlation between position angles and other quantity, such as the direction of Gaia DR2 proper motion, direction of major axis of error ellipses;

This notebook show the first part of this work, where I did NOT remove the global systematical difference amongst 
catalogs before my analysis.

## 1. Multi-wavelength positional offsets
### 1.1 Calculation

```python
from astropy.table import Table, join, Column
from astropy import units as u
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, NullFormatter
%matplotlib inline
```

Get the quasar positions from 4 catalogs and cross-match them to find the common sources.

```python
from my_progs.catalog import read_icrf, read_gaia

icrf3sx = read_icrf.read_icrf3(wv="sx")
icrf3k = read_icrf.read_icrf3(wv="k")
icrf3xka = read_icrf.read_icrf3(wv="xka")

gaiadr2 = read_gaia.read_dr2_iers()
```

```python
tmp1 = join(icrf3sx, icrf3k, keys=["icrf_name", "iers_name"], table_names=["sx", "k"])
tmp2 = join(icrf3xka, gaiadr2, keys="iers_name", table_names=["xka", "opt"])

comsou = join(tmp1, tmp2, keys=["icrf_name", "iers_name"])

print("We obtain {:d} common sources amongst 4 catalogs.".format(len(comsou)))
```

```python
from my_progs.catalog.pos_diff import radio_cat_diff_calc
# Calculate the positional offset wrt. S/X position
# K - S/X
k2sx = radio_cat_diff_calc(icrf3k, icrf3sx, "iers_name", ["k", "sx"])

# X/Ka - S/X
xka2sx = radio_cat_diff_calc(icrf3xka, icrf3sx, "iers_name", ["xka", "sx"])
xka2sx.remove_columns(['ra', 'dec', 'ra_err_sx', 'dec_err_sx', 'pos_err_sx'])

# Optical - S/X
opt2sx = radio_cat_diff_calc(gaiadr2, icrf3sx, "iers_name", ["opt", "sx"])
opt2sx.remove_columns(['ra', 'dec', 'ra_err_sx', 'dec_err_sx', 'pos_err_sx'])
```

```python
comsou = join(k2sx, xka2sx, keys="iers_name", table_names=["k", "xka"])
comsou = join(comsou, opt2sx, keys="iers_name")

# Rename some columns
comsou.rename_columns(["dra", "ddec", "dra_err", "ddec_err", "dra_ddec_cov",
                       "ang_sep", "nor_ra", "nor_dec", "nor_sep"],
                      ["dra_opt", "ddec_opt", "dra_err_opt", "ddec_err_opt",
                       "dra_ddec_cov_opt",
                       "ang_sep_opt", "nor_ra_opt", "nor_dec_opt", "nor_sep_opt"])


# Add proper motion information
gdr2pm = Table(gaiadr2)
gdr2pm.keep_columns(["iers_name", "pmra", "pmra_err", "pmdec", "pmdec_err", "ra_dec_corr"])
gdr2pm.rename_column("ra_dec_corr", "ra_dec_corr_opt")

comsou = join(comsou, gdr2pm, keys="iers_name")

# Add correlation coefficent
corrsx = Table(icrf3sx)
corrsx.keep_columns(["iers_name", "ra_dec_corr"])
corrsx.rename_column("ra_dec_corr", "ra_dec_corr_sx")

corrk = Table(icrf3k)
corrk.keep_columns(["iers_name", "ra_dec_corr"])
corrk.rename_column("ra_dec_corr", "ra_dec_corr_k")

corrxka = Table(icrf3xka)
corrxka.keep_columns(["iers_name", "ra_dec_corr"])
corrxka.rename_column("ra_dec_corr", "ra_dec_corr_xka")

comsou = join(comsou, corrsx, keys="iers_name")
comsou = join(comsou, corrk, keys="iers_name")
comsou = join(comsou, corrxka, keys="iers_name")
```

```python
comsou.write("../logs/common-soource-position.fits", format="fits", overwrite=True)
```

There are 488 common sources in the four catalogs.

I use the S/X-band position as the reference point and calculate the positional offset vector of K-band, X/Ka-band, and Gaia, 
as illustrated in the following diagram.

Please note in the plot $\Delta\alpha\cos\delta$ decreases from the left to right.
This kind of convention is used in the VLBI imaging so I follow the same convention for further comparison.

From the position offset vector I calculate the angular separation $\rho$ 
and normalized separation $X$ as done in the Mignard et 2018. 

```python
# Positional offset vector
fig, ax = plt.subplots()

ax.plot(0, 0, "k+", label="S/X", ms=10)
ax.plot(comsou[0]["dra_k"], comsou[0]["ddec_k"], "rs", label="K")
ax.plot(comsou[0]["dra_xka"], comsou[0]["ddec_xka"], "bv", label="X/Ka")
ax.plot(comsou[0]["dra_opt"], comsou[0]["ddec_opt"], "g^", label="Gaia")

ax.arrow(0, 0, comsou[0]["dra_k"], comsou[0]["ddec_k"], color="r")
ax.arrow(0, 0, comsou[0]["dra_xka"], comsou[0]["ddec_xka"], color="b")
ax.arrow(0, 0, comsou[0]["dra_opt"], comsou[0]["ddec_opt"], color="g")

### TO DO 
### Plot error ellipse

ax.set_xlim(0.5, -0.2)

ax.set_xlabel("$\Delta\,\\alpha\,\cos\delta$ (mas)")
ax.set_ylabel("$\Delta\,\delta$ (mas)")
ax.set_title(comsou[0]["iers_name"])

ax.grid()
ax.legend()
```

## Positional angle of position offset vector

The position angle, which characterize direction of position offset vector, is measured as angle between the direction of increasing-declination and position offset vector clockwisely.

As a result, the position angles of K-band, X/Ka-band, and Gaia position wrt. the S/X band one for 0002+541 are $217^\circ$, $135^\circ$, $63^\circ$, respectively.

```python
def pa_calc(dra, ddec):
    """Calculate positional angle from positional offset.

    Parametes
    ---------
    dra : ndarray
        positional difference in R.A.(times cos(decl.))
    ddec : ndarray
        positional difference in declination

    Returns
    -------
    A : ndarray
        Angle (in degree) of positional offset vector towards to y-axis clockwisely
    """

    # Direction of positional offset vector from the y-axis count-clockwisely
    # in degree
    A = np.rad2deg(np.arctan2(dra, ddec))
    
    A = np.where(A>0, A, A+360)

    return A
```

```python
# Calculate the positional angle of the positional offset
PA_k = pa_calc(comsou["dra_k"], comsou["ddec_k"])
PA_xka = pa_calc(comsou["dra_xka"], comsou["ddec_xka"])
PA_opt = pa_calc(comsou["dra_opt"], comsou["ddec_opt"])
```

The diagram below despicts the distribution of positional angles for K-band, X/Ka-band, and Gaia position wrt. S/X-band position.

A steep peak aroud $180^\circ$ is seen in X/Ka$-$S/X, most likely caused by the negitive declination bias seen in plots above.

There is no favorable direction of K$-$S/X and Gaia$-$S/X.

In the following figure, I directly look at the angle between positional offset vectors.

```python
def inclu_angle_calc(dra1, ddec1, dra2, ddec2):
    """Calculate included angle between two positional offset vectors.


    Parametes
    ---------
    dra1, ddec1: ndarray
        components of the fisrt positional offset vector
    dra2, ddec2: ndarray
        components of the second positional offset vectors

    Returns
    -------
    ang : ndarray
        Angle (in degree) between positional offset vectors
    """

    cosine = dra1 * dra2 + ddec1 * ddec2
    mod1 = np.sqrt(dra1**2 + ddec1**2)
    mod2 = np.sqrt(dra2**2 + ddec2**2)

    ang = np.rad2deg(np.arccos(cosine/mod1/mod2))

    return ang
```

```python
# Calculate the angle between positional offset vectors
ang_k_opt = inclu_angle_calc(
    comsou["dra_k"], comsou["ddec_k"], comsou["dra_opt"], comsou["ddec_opt"])
ang_xka_opt = inclu_angle_calc(
    comsou["dra_xka"], comsou["ddec_xka"], comsou["dra_opt"], comsou["ddec_opt"])
ang_xka_k = inclu_angle_calc(
    comsou["dra_k"], comsou["ddec_k"], comsou["dra_xka"], comsou["ddec_xka"])
```

To check if the position angle of positional offset vector is biased by the origin position formal error at each band, I calculate the position angle of the direction of major axis of error ellipse (EEMA) for each catalog.


```python
def eepa_calc_single(ra_err, dec_err, ra_dec_corr):
    """Calculate the ovrall formal uncertainty.

    ovrall formal uncertainty = sqrt(RA_err^2+Dec_err^2+C*RA_err*Dec_err)

    Parameters
    ----------
    ra_err/dec_err : formal uncertainty of RA/Dec, usually in micro-as
    ra_dec_corr : correlation coeffient between RA and Dec, unitless.

    Returns
    ----------
    pa : the position angle of the major axis of the error ellipse
    """

    from numpy.linalg import eig
    
    cov = ra_dec_corr * ra_err * dec_err
    cov_mat = np.array([[ra_err**2, cov], [cov, dec_err**2]])
    
    eig_val, eig_vec = eig(cov_mat)

    if eig_val[0] > eig_val[1]:
        vec_M = eig_vec[:, 0]
    else:
        vec_M = eig_vec[:, 1]

    # Ensure that eigenvector locates in the 1st or 4th quadrant
    if vec_M[0] < 0:
        vec_M = -vec_M

    pa0 = np.rad2deg(np.arctan2(vec_M[1], vec_M[0]))

    if pa0 <= 90:
        pa = 90 - pa0
    else:
        pa = 450 - pa0

    return pa

def eepa_calc(ra_err, dec_err, ra_dec_corr):
    """Calculate the ovrall formal uncertainty.

    ovrall formal uncertainty = sqrt(RA_err^2+Dec_err^2+C*RA_err*Dec_err)

    Parameters
    ----------
    ra_err/dec_err : formal uncertainty of RA/Dec, usually in micro-as
    ra_dec_corr : correlation coeffient between RA and Dec, unitless.

    Returns
    ----------
    pa : the position angle of the major axis of the error ellipse
    """

    pa = [eepa_calc_single(ra_erri, dec_erri, ra_dec_corri)
          for ra_erri, dec_erri, ra_dec_corri in
          zip(ra_err, dec_err, ra_dec_corr)]

    # Turn list to np.array
    pa = np.array(pa)

    return pa
```

```python
# Calculate the positional angle of the major axis of error ellipse
eepa_sx = eepa_calc(comsou["ra_err_sx"], comsou["dec_err_sx"], comsou["ra_dec_corr_sx"])
eepa_k = eepa_calc(comsou["ra_err_k"], comsou["dec_err_k"], comsou["ra_dec_corr_k"])
eepa_xka = eepa_calc(comsou["ra_err_xka"], comsou["dec_err_xka"], comsou["ra_dec_corr_xka"])
eepa_opt = eepa_calc(comsou["ra_err_opt"], comsou["dec_err_opt"], comsou["ra_dec_corr_opt"])
```

The error ellipse from VLBI catalogs (S/X, K, and X/Ka) has a priviledge direction of $\sim180^\circ$, especially for K-band, which can be expected.

```python
# Histogram of position angle of eema
fig, ax = plt.subplots(figsize=(8, 6))

bins_set = np.linspace(0, 180, 25)

# Weights to calculate the per cent
wgt = 100 / len(comsou) * np.ones_like(comsou["ra"])

# Plot
ax.hist([eepa_sx, eepa_k, eepa_xka, eepa_opt],
        bins_set, histtype="step", fill=False,
        color=["k", "r", "b", "g"],
        label=["S/X", "K", "X/Ka", "Gaia"],
        weights=[wgt, wgt, wgt, wgt])

ax.set_xticks(np.arange(0, 181, 30))
ax.set_ylabel("%", fontsize=12)
ax.set_xlabel("Position angle of EEMA [$^\circ$]", fontsize=12)
ax.xaxis.set_major_locator(MultipleLocator(10))

ax.legend(loc="upper left")
# ax.grid()
ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")
```
