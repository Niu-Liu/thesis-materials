In this notebook, I comparison the several ICRF3-prototype solutions with the Gaia DR2 catalog.

```python
from astropy.table import Table, join, Column
from astropy import units as u
import astropy.coordinates as coord
from astropy.coordinates import SkyCoord
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# My modules
from my_progs.catalog.vsh_deg1_cor import vsh_deg01_fitting
from my_progs.catalog.pos_diff import pos_diff_calc
from my_progs.catalog.read_gaia import read_dr2_iers
from my_progs.catalog.glide_calc import glide_calc, GA_glide_decomposed
from my_progs.vlbi.read_sou import read_sou
```

First step is to read Gaia DR2 catalog.

```python
gaiadr2 = read_dr2_iers()
```

Then read the source position from the VLBI solutions.

```python
x1 = np.arange(3.0, 5.0, 0.5)
x2 = np.array([5.0, 5.1, 5.3, 5.4, 5.5, 5.6, 5.7, 5.8, 5.9])
x3 = np.arange(6.0, 10.0, 0.5)

GAs = np.concatenate((x1, x2, x3))
```

```python
data_dir = "/Users/Neo/Astronomy/data/vlbi_solutions/vlbi2_server/galactic-aberration"

d1 = np.zeros_like(GAs)
d2 = np.zeros_like(GAs)
d3 = np.zeros_like(GAs)
d = np.zeros_like(GAs)
d1_err = np.zeros_like(GAs)
d2_err = np.zeros_like(GAs)
d3_err = np.zeros_like(GAs)
d_err = np.zeros_like(GAs)

d_gc = np.zeros_like(GAs)
d_gc_err = np.zeros_like(GAs)
d_ngc = np.zeros_like(GAs)
d_ngc_err = np.zeros_like(GAs)
```

```python
for i, GA in enumerate(GAs):
    t_sol = read_sou("%s/GA%3.1f_15.5/GA%3.1f_15.5.sou" % (data_dir, GA, GA))
    
    # Corss-match
    com_sou = join(t_sol, gaiadr2, keys="iers_name", table_names=["opa", "gaiadr2"])
    
    [dRA, dDC, dRA_err, dDC_err, dRA_dDC_cov,
     ang_sep, X_a, X_d, X] = pos_diff_calc(
        com_sou["ra_opa"], com_sou["ra_err_opa"], 
        com_sou["dec_opa"], com_sou["dec_err_opa"], 
        com_sou["ra_dec_corr_opa"],
        com_sou["ra_gaiadr2"], com_sou["ra_err_gaiadr2"],
        com_sou["dec_gaiadr2"], com_sou["dec_err_gaiadr2"], 
        com_sou["ra_dec_corr_gaiadr2"])

    com_sou.add_columns([dRA, dDC, dRA_err, dDC_err, dRA_dDC_cov, ang_sep, X_a, X_d, X],
                        names=["dra", "ddec", "dra_err", "ddec_err", "dra_ddec_cov",
                               "ang_sep", "nor_dra",  "nor_ddec", "nor_sep"])
    
    # Remove the outlier (consider the normalized separation)
    N = len(com_sou)
    X0 = np.sqrt(np.log(N))

    print("There are %d sources.\n"
          "When X >= %.2f, the number of the outlier is smaller than 1." % (N, X0))

    mask = ((com_sou["nor_sep"] <= X0)
            & (com_sou["ang_sep"] < 10))

    # Table of a clean sample
    com_clean = com_sou[mask]

    print("After elimination, there are %d sources in the clean sample." % len(com_clean))
    

    # Transform columns into np.array
    dra = np.array(com_clean["dra"])
    ddec = np.array(com_clean["ddec"])
    dra_err = np.array(com_clean["dra_err"])
    ddec_err = np.array(com_clean["ddec_err"])
    ra_rad = np.array(com_clean["ra_opa"].to(u.radian))
    dec_rad = np.array(com_clean["dec_opa"].to(u.radian))
    dra_ddec_cov = np.array(com_clean["dra_ddec_cov"])
    
    # Transformation parameters
    # l_max = 1
    w2_cln, sig2_cln, corrcoef2_cln, _, _, _ = vsh_deg01_fitting(
        dra, ddec, ra_rad, dec_rad, dra_err, ddec_err,
        cov=dra_ddec_cov, elim_flag="None")

    # mas -> uas
    w2 = w2_cln * 1.e3
    sig2 = sig2_cln * 1.e3


    d1[i], d1_err[i] = w2[0], sig2[0]
    d2[i], d2_err[i] = w2[1], sig2[1]
    d3[i], d3_err[i] = w2[2], sig2[2]

    # Decomposed the glide vector
    A, _, _, A_err, _, _ = glide_calc(w2[:3], sig2[:3])
    D_GA, D_GA_err, D_nGA, _, _, D_nGA_err,  _, _ = GA_glide_decomposed(w2[:3], sig2[:3])
    
    d[i], d_err[i] = A, A_err
    d_gc[i], d_gc_err[i] = D_GA, D_GA_err
    d_ngc[i], d_ngc_err[i] = D_nGA, D_nGA_err
```

```python
fig, ax = plt.subplots()
ax.errorbar(GAs, d1, d1_err, marker="*", elinewidth=1, label="$g_1$")
ax.errorbar(GAs, d2, d2_err, marker="^", elinewidth=1, label="$g_2$")
ax.errorbar(GAs, d3, d3_err, marker="s", elinewidth=1, label="$g_3$")

# ax.set_ylabel("Glide / ($\mathrm{\mu as}$)")
# ax.set_xlabel("GA / ($\mathrm{\mu as \cdot yr^{-1}}$)")
ax.set_ylabel("Glide / $\mathrm{\mu as}$")
ax.set_xlabel("GA / $\mathrm{\mu as \cdot yr^{-1}}$")

ax.legend()

ax.set_ylim([-30, 35])

ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

# plt.savefig("/Users/Neo/Desktop/glide.png")
# plt.savefig("/Users/Neo/Desktop/glide1.eps")
```
