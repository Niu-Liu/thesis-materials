```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# File name: catalog_error_comparison.py
"""
Created on Sun May 27 15:28:04 2018

@author: Neo(liuniu@smail.nju.edu.cn)

Plot the smoothed formal error as a function of declination of various catalogs.

"""

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from matplotlib import pyplot as plt
import numpy as np

# My modules
from my_progs.catalog.read_icrfn import read_icrf1, read_icrf2, read_icrf3
from my_progs.catalog.read_gaia import read_dr1_qso, read_dr2_qso
```

Read the information from catalogs.

```python
# ICRF1 catalog
icrf1 = read_icrf1()
icrf1.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])
icrf1 = icrf1.filled()

# ICRF2 catalog
icrf2 = read_icrf2()
icrf2.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])

# ICRF3 S/X catalog
icrf3sx = read_icrf3(wv="sx")
icrf3sx.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])

# ICRF3 K catalog
icrf3k = read_icrf3(wv="k")
icrf3k.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])

# ICRF3 X/Ka catalog
icrf3xka = read_icrf3(wv="xka")
icrf3xka.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])

# Gaia DR1 Quasar auxiliary solution
gdr1 = read_dr1_qso()
gdr1.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])

# Gaia DR2 aux_iers catalog
gdr2 = read_dr2_qso()
gdr2.keep_columns(["ra", "ra_err", "dec_err", "pos_err"])
```

I bin the radio sources by their declination with a 5-degree interval.

```python
# Bin the data by 5 degree interval

# ICRF1
icrf1_ra_bin = np.trunc(icrf1["ra"] / 5)

icrf1_grouped = icrf1.group_by(icrf1_ra_bin)
icrf1_binned = icrf1_grouped.groups.aggregate(np.median)

# ICRF2
icrf2_ra_bin = np.trunc(icrf2["ra"] / 5)

icrf2_grouped = icrf2.group_by(icrf2_ra_bin)
icrf2_binned = icrf2_grouped.groups.aggregate(np.median)

# ICRF3 SX
icrf3sx_ra_bin = np.trunc(icrf3sx["ra"] / 5)

icrf3sx_grouped = icrf3sx.group_by(icrf3sx_ra_bin)
icrf3sx_binned = icrf3sx_grouped.groups.aggregate(np.median)

# ICRF3 K
icrf3k_ra_bin = np.trunc(icrf3k["ra"] / 5)

icrf3k_grouped = icrf3k.group_by(icrf3k_ra_bin)
icrf3k_binned = icrf3k_grouped.groups.aggregate(np.median)

# ICRF3 XKa
icrf3xka_ra_bin = np.trunc(icrf3xka["ra"] / 5)

icrf3xka_grouped = icrf3xka.group_by(icrf3xka_ra_bin)
icrf3xka_binned = icrf3xka_grouped.groups.aggregate(np.median)

# Gaia DR1
gdr1_ra_bin = np.trunc(gdr1["ra"] / 5)

gdr1_grouped = gdr1.group_by(gdr1_ra_bin)
gdr1_binned = gdr1_grouped.groups.aggregate(np.median)

# Gaia DR2
gdr2_ra_bin = np.trunc(gdr2["ra"] / 5)

gdr2_grouped = gdr2.group_by(gdr2_ra_bin)
gdr2_binned = gdr2_grouped.groups.aggregate(np.median)
```

```python
# Bin the data by 5 degree interval

binsize = 20

# ICRF1
icrf1_ra_bin = np.trunc(np.arange(len(icrf1)) / binsize)
icrf1.sort("ra")
icrf1_grouped = icrf1.group_by(icrf1_ra_bin)
icrf1_binned = icrf1_grouped.groups.aggregate(np.median)

binsize = 50

# ICRF2
icrf2_ra_bin = np.trunc(np.arange(len(icrf2)) / binsize)
icrf2.sort("ra")
icrf2_grouped = icrf2.group_by(icrf2_ra_bin)
icrf2_binned = icrf2_grouped.groups.aggregate(np.median)

# ICRF3 SX
icrf3sx_ra_bin = np.trunc(np.arange(len(icrf3sx)) / binsize)
icrf3sx.sort("ra")
icrf3sx_grouped = icrf3sx.group_by(icrf3sx_ra_bin)
icrf3sx_binned = icrf3sx_grouped.groups.aggregate(np.median)

binsize = 20

# ICRF3 K
icrf3k_ra_bin = np.trunc(np.arange(len(icrf3k)) / binsize)
icrf3k.sort("ra")
icrf3k_grouped = icrf3k.group_by(icrf3k_ra_bin)
icrf3k_binned = icrf3k_grouped.groups.aggregate(np.median)

# ICRF3 XKa
icrf3xka_ra_bin = np.trunc(np.arange(len(icrf3xka)) / binsize)
icrf3xka.sort("ra")
icrf3xka_grouped = icrf3xka.group_by(icrf3xka_ra_bin)
icrf3xka_binned = icrf3xka_grouped.groups.aggregate(np.median)

binsize = 50

# Gaia DR1
gdr1_ra_bin = np.trunc(np.arange(len(gdr1)) / binsize)
gdr1.sort("ra")
gdr1_grouped = gdr1.group_by(gdr1_ra_bin)
gdr1_binned = gdr1_grouped.groups.aggregate(np.median)

# Gaia DR2
gdr2_ra_bin = np.trunc(np.arange(len(gdr2)) / binsize)
gdr2.sort("ra")
gdr2_grouped = gdr2.group_by(gdr2_ra_bin)
gdr2_binned = gdr2_grouped.groups.aggregate(np.median)
```

```python
# Plot for median error
fig, (ax0, ax1) = plt.subplots(figsize=(10, 4), ncols=2)

# Right ascension
ax0.plot(gdr2_binned["ra"], gdr2_binned["ra_err"],
         "k-^", ms=5, label="Gaia DR2")
ax0.plot(icrf1_binned["ra"], icrf1_binned["ra_err"],
         "g-v", ms=5, label="ICRF1")
ax0.plot(icrf2_binned["ra"], icrf2_binned["ra_err"],
         "m-s", ms=5, label="ICRF2")
ax0.plot(icrf3sx_binned["ra"], icrf3sx_binned["ra_err"],
         "r-x", ms=5, label="ICRF3 SX")
ax0.plot(icrf3k_binned["ra"], icrf3k_binned["ra_err"],
         "y-o", ms=5, label="ICRF3 K")
ax0.plot(icrf3xka_binned["ra"], icrf3xka_binned["ra_err"],
         "b-*", ms=5, label="ICRF3 XKa")

ax0.set_xlim([0, 360])
ax0.set_ylim([0.04, 3.5])
ax0.set_xticks(np.arange(0, 361, 60))
ax0.set_xlabel("Right ascension")
ax0.set_yscale("log")
ax0.set_ylabel("Formal Error in $\\alpha^*$ (mas)")
ax0.grid()
ax0.legend(fontsize="x-small", loc="upper right")

ax0.xaxis.set_ticks_position("both")
ax0.yaxis.set_ticks_position("both")

# Declination
ax1.plot(gdr2_binned["ra"], gdr2_binned["dec_err"],
         "k-^", ms=5, label="Gaia DR2")
ax1.plot(icrf1_binned["ra"], icrf1_binned["dec_err"],
         "g-v", ms=5, label="ICRF1")
ax1.plot(icrf2_binned["ra"], icrf2_binned["dec_err"],
         "m-s", ms=5, label="ICRF2")
ax1.plot(icrf3sx_binned["ra"], icrf3sx_binned["dec_err"],
         "r-x", ms=5, label="ICRF3 SX")
ax1.plot(icrf3k_binned["ra"], icrf3k_binned["dec_err"],
         "y-o", ms=5, label="ICRF3 K")
ax1.plot(icrf3xka_binned["ra"], icrf3xka_binned["dec_err"],
         "b-*", ms=5, label="ICRF3 XKa")

ax1.set_xlim([0, 360])
ax1.set_ylim([0.04, 4])
ax1.set_xticks(np.arange(0, 361, 60))
ax1.set_xlabel("Right ascension")
ax1.set_yscale("log")
ax1.set_ylabel("Formal Error in $\\delta$ (mas)")
ax1.grid()
ax1.legend(fontsize="x-small", loc="upper right")

ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")

plt.tight_layout()
# plt.subplots_adjust(hspace=0.01)

plt.savefig("../plots/pos-err-vs-ra.eps")
```
