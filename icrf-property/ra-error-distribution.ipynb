I plot the spatial distribution of formal error in the right ascension of ICRF catalogs and Gaia DR2.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from astropy.coordinates import SkyCoord
from matplotlib import pyplot as plt, cm as cm
import numpy as np

from astropy import units as u
import numpy as np
from numpy import cos, deg2rad, sqrt, pi

# My modules
from my_progs.catalog.read_icrfn import read_icrf1, read_icrf2, read_icrf3
from my_progs.catalog.read_gaia import read_dr2_qso
```

Read catalogs.

```python
# ICRF1 catalog
icrf1 = read_icrf1()
icrf1.keep_columns(["ra", "dec", "ra_err", "dec_err", "pos_err", "ra_dec_corr"])

# Fill the masked column
pos_err = icrf1["pos_err"].filled()

# ICRF2 catalog
icrf2 = read_icrf2()
icrf2.keep_columns(["ra", "dec", "ra_err", "dec_err", "pos_err", "ra_dec_corr"])

# ICRF3 S/X catalog
icrf3sx = read_icrf3(wv="sx")
icrf3sx.keep_columns(["ra", "dec", "ra_err", "dec_err", "pos_err", "ra_dec_corr"])

# ICRF3 K catalog
icrf3k = read_icrf3(wv="k")
icrf3k.keep_columns(["ra", "dec", "ra_err", "dec_err", "pos_err", "ra_dec_corr"])

# ICRF3 X/Ka catalog
icrf3xka = read_icrf3(wv="xka")
icrf3xka.keep_columns(["ra", "dec", "ra_err", "dec_err", "pos_err", "ra_dec_corr"])

# Gaia DR2 aux_iers catalog
gaiadr2 = read_dr2_qso()
gaiadr2.keep_columns(["ra", "dec", "ra_err", "dec_err", "pos_err", "ra_dec_corr"])
```

```python
# Transform the coordinate into an astropy.coordinate.SkyCoord object
# ## opa solution
# coord_v = SkyCoord(ra=ra_com_v*u.deg, dec=dec_com_v*u.deg)
# # The mid-point of skyplot will be 0
# ra_rad_v = coord_v.ra.wrap_at(180 * u.deg).radian
# dec_rad_v = coord_v.dec.radian

## Gaia DR2
coord_g = SkyCoord(ra=gaiadr2["ra"], dec=gaiadr2["dec"])
ra_rad_g = coord_g.ra.wrap_at(180 * u.deg).radian
dec_rad_g = coord_g.dec.radian
```

```python
# Plot the formal error distribution of opa and Gaia DR2
fig = plt.figure(figsize=(8, 4))

# GaiaDR2 - RA_err
ax3 = plt.subplot(111, projection="aitoff", facecolor="lightcyan")
sc3 = ax3.scatter(ra_rad_g, dec_rad_g, marker=".", s=5,
                  c=gaiadr2["ra_err"], vmin=0, vmax=0.5, cmap="rainbow")
ax3.set_xticks(np.arange(-5./6 * pi, pi, pi/6))
ax3.set_xticklabels(["14h", "16h", "18h", "20h", "22h",
                     "0h", "2h", "4h", "6h", "8h", "10h"])
plt.colorbar(sc3,
             extend="max", label="${\it Gaia}$ DR2 (mas)")
ax3.grid(True)

# galactic plane
lon = np.arange(360)
lat = np.zeros_like(lon)
coord_ecl = SkyCoord(lon=lon*u.deg, lat=lat*u.deg,
                     frame="barycentrictrueecliptic")

# The mid-point of skyplot will be 0
ra_rad_ecl = coord_ecl.icrs.ra.wrap_at(180 * u.deg).radian
dec_rad_ecl = coord_ecl.icrs.dec.radian
ax3.plot(ra_rad_ecl, dec_rad_ecl, "k.", ms=0.3)

plt.tight_layout()
```
