In this notebook, I look at the number of observations in the ICRF and $\it Gaia$ catalogs.

```python
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = "svg"
# My modules
from my_progs.catalog.read_icrfn import read_icrf1, read_icrf2, read_icrf3
```

```python
def simple_stat(x):
    
    mean = np.mean(x)
    median = np.median(x)
    maximum = np.max(x)
    minimum = np.min(x)
    
    print("The maximum is {:10.0f}".format(maximum))
    print("    minimum is {:10.0f}".format(minimum))
    print("    mean    is {:10.0f}".format(mean))
    print("    median  is {:10.0f}".format(median))
```

```python
# ICRF1 catalog
icrf1 = read_icrf1()

simple_stat(icrf1["nb_del"])

fig, ax = plt.subplots()

ax.hist(icrf1["nb_del"], 30, log=True)
```

```python
# ICRF2 catalog
icrf2 = read_icrf2()

simple_stat(icrf2["nb_del"])

plt.hist(icrf2["nb_del"], 30, log=True)
```

```python
icrf3sx = read_icrf3(wv="sx")

simple_stat(icrf3sx["nb_del"])

plt.hist(icrf3sx["nb_del"], 30, log=True)
```
