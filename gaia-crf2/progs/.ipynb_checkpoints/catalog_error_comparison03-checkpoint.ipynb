```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# File name: catalog_error_comparison.py
"""
Created on Sun May 27 15:28:04 2018

@author: Neo(liuniu@smail.nju.edu.cn)

Plot the smoothed declination difference as a function of declination of various catalogs.

"""

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from matplotlib import pyplot as plt
import numpy as np
# My modules
from read_icrf1 import read_icrf1_pos
from read_icrf2 import read_icrf2
from read_sou import read_cat
from read_GaiaDR1 import read_gaiadr1
from read_GaiaDR2 import read_gaiadr2_iers_position
from cross_match import (list_crossmatch, postional_difference_calc)
```

```python
# Read catalog data
# ICRF1 catalog
[icrf_name_i1, iers_name_i1,
 ra_i1, dec_i1, ra_error_i1, dec_error_i1, ra_dec_corr_i1, _] = read_icrf1_pos("../data/rsc95r01.dat")

# ICRF2 catalog
[icrf_name_i2, ivs_name_i2, iers_name_i2,
 ra_i2, dec_i2, ra_error_i2, dec_error_i2, ra_dec_corr_i2, _, _] = read_icrf2("../data/icrf2.dat")

# VLBI solution OPAA
vlbi_cat_v1 = "../data/opa-sx-180425-noGA.cat"
[ivs_name_v1, iers_name_v1, ra_v1, dec_v1, ra_error_v1, dec_error_v1,
 ra_dec_corr_v1, num_ses_v1, num_obs_v1] = read_cat(vlbi_cat_v1)

# VLBI solution OPAB
[ivs_name_v2, iers_name_v2, ra_v2, dec_v2, ra_error_v2, dec_error_v2,
 ra_dec_corr_v2, num_ses_v2, num_obs_v2] = read_cat("../data/opa-sx-180425-GA00.cat")

# VLBI solution OPAC
[ivs_name_v3, iers_name_v3, ra_v3, dec_v3, ra_error_v3, dec_error_v3,
 ra_dec_corr_v3, num_ses_v3, num_obs_v3] = read_cat("../data/opa-sx-180425-GA15.cat")

# Gaia DR1 data
[icrf_name_g1, ra_g1, dec_g1, 
 ra_error_g1, dec_error_g1, ra_dec_corr_g1, _] = read_gaiadr1("../data/qso.dat")

# Gaia DR2 data
[iers_name_g2, ra_g2, ra_error_g2,
 dec_g2, dec_error_g2, ra_dec_corr_g2] = read_gaiadr2_iers_position("../data/gaiadr2_iers.fits")
```

```python
def dec_smoothed_dec_diff(dec, ddec, interv_size=50):
    """Calculate the smoothed declination difference as a function of declination

    Parameters
    ----------
    dec : array_like of float type
        declination in degree
    ddec : array_like of float type
        difference in declination in mas
    interv_size : int
        number of source in a subset, default is 50.

    Returns
    ----------
    med_dec : array of float
        median declination of every subset
    med_ddec : array of int
        median declination difference of every subset
    """

    # Sort the data according to the declination
    ind = np.argsort(dec)
    dec_sort = np.take(dec, ind)
    ddec_sort = np.take(ddec, ind)

    # Then we calculate the median ellpise major axis for subset of 50 sources obtained by a moving slide windows.
    if dec_sort.size <= interv_size:
        print("# Too smaller sample!")
        exit()
    else:
        interv_num = dec_sort.size - interv_size + 1

    med_dec = np.zeros(interv_num)
    med_ddec = np.zeros(interv_num)

    for i in range(interv_num):
        ind_b, ind_e = i, i + interv_size
        deci = dec_sort[ind_b: ind_e]
        ddeci = ddec_sort[ind_b: ind_e]

        med_dec[i] = np.median(deci)
        med_ddec[i] = np.median(ddeci)

    return med_dec, med_ddec
```

```python
def dec_diff_calc(iers_name1, dec1, iers_name2, dec2):
    """Calculate the declination for common sources between two catalogs.
    
    Parameters
    ----------
    iers_name1/iers_name2 : array_like of string
        IERS designation source name in two catalogs
    dec1/dec2 : array_like of float
        Declination of two catalogs. Units are degree.
    
    Returns
    ---------
    dec_com : array_like of float
        Declination of common sources between two catalogs. 
        Units are degree.
    ddec : array_like of float
        Declination difference of common sources between two catalogs.
        Units are mas.
    """

    # Cross-match between between two catalogs
    comsou_iers, ind1, ind2 = list_crossmatch(iers_name1, iers_name2)
    print("There are %d common source between catalog 1 (%d sources) and catalog 2 (%d sources)" 
          %(comsou_iers.size, iers_name1.size, iers_name2.size))

    # Verify the result of cross-match
    soucom1 = np.take(iers_name1, ind1)
    soucom2 = np.take(iers_name2, ind2)
    for i, (comsoui, soucom1i, soucom2i) in enumerate(
            zip(comsou_iers, soucom1, soucom2)):

        if comsoui != soucom1i:
            print("%dth source %s are not consistented in catalog 1 %s." %
                  (i, comsoui, soucom1i))

        if comsoui != soucom2i:
            print("%dth source %s are not consistented in catalog 2 %s." %
                  (i, comsoui, soucom2i))


    # Extract data for common source
    # Catalog1 data
    dec_com1 = np.take(dec1, ind1)

    # Catalog2 data
    dec_com2 = np.take(dec2, ind2)


    # compute the delination difference
    dec = dec_com1
    ddec = (dec_com1 - dec_com2) * 3.6e6
    
    return dec, ddec
```

```python
# Use GaiaDR2 position as the reference, calculate the declination difference

# ICRF1 - GaiaDR2
dec_i1g2, ddec_i1g2 = dec_diff_calc(iers_name_i1, dec_i1, iers_name_g2, dec_g2)

# ICRF2 - GaiaDR2
dec_i2g2, ddec_i2g2 = dec_diff_calc(iers_name_i2, dec_i2, iers_name_g2, dec_g2)

# OPAA - GaiaDR2
dec_v1g2, ddec_v1g2 = dec_diff_calc(iers_name_v1, dec_v1, iers_name_g2, dec_g2)

# OPAB - GaiaDR2
dec_v2g2, ddec_v2g2 = dec_diff_calc(iers_name_v2, dec_v2, iers_name_g2, dec_g2)

# OPAC - GaiaDR2
dec_v3g2, ddec_v3g2 = dec_diff_calc(iers_name_v3, dec_v3, iers_name_g2, dec_g2)
```

```python
# Calculate the smoothed declination difference.

bin_size = 50

# ICRF1 - GaiaDR2
med_dec_i1g2, med_ddec_i1g2 = dec_smoothed_dec_diff(dec_i1g2, ddec_i1g2, bin_size)

# ICRF2 - GaiaDR2
med_dec_i2g2, med_ddec_i2g2 = dec_smoothed_dec_diff(dec_i2g2, ddec_i2g2, bin_size)

# OPAA - GaiaDR2
med_dec_v1g2, med_ddec_v1g2 = dec_smoothed_dec_diff(dec_v1g2, ddec_v1g2, bin_size)

# OPAB - GaiaDR2
med_dec_v2g2, med_ddec_v2g2 = dec_smoothed_dec_diff(dec_v2g2, ddec_v2g2, bin_size)

# OPAC - GaiaDR2
med_dec_v3g2, med_ddec_v3g2 = dec_smoothed_dec_diff(dec_v3g2, ddec_v3g2, bin_size)
```

```python
# Plot for median error
fig, ax = plt.subplots()


ax.plot(med_dec_i1g2, med_ddec_i1g2, "b.", ms=2, label="ICRF1")
ax.plot(med_dec_i2g2, med_ddec_i2g2, "k.", ms=2, label="ICRF2")
ax.plot(med_dec_v1g2, med_ddec_v1g2, "r.", ms=2, label="OPAA")
ax.plot(med_dec_v2g2, med_ddec_v2g2, "y.", ms=2, label="OPAB")
ax.plot(med_dec_v3g2, med_ddec_v3g2, "g.", ms=2, label="OPAC")

ax.set_xlim([-90, 90])
ax.set_xticks(np.arange(-90, 91, 30))
ax.set_xlabel("Declination (degree)")
ax.set_ylabel('Error (mas)')
ax.grid()  
ax.legend(fontsize="small")
```
