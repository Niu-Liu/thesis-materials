```python
from astropy.table import Table
from astropy.io import fits
```

```python
# table = Table("../data/1536347799833O-result.fits")
fits_filename = "../data/0134+329-sourrouding.fits"
hdul = fits.open(fits_filename)
```

```python
data = hdul[1].data
ra = data["ra"]
dec = data["dec"]
```

```python
# VLBI position for 0134+329
ra0 = 24.422084894
dec0 = 33.1597655334

# Gaia DR2 position for 0134+329
ra1 = 24.422081768796943
dec1 = 33.159744439930684

dra = (ra - ra0) * 3.6e3
ddec = (dec - dec0) * 3.6e3
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.format = "svg"
plt.scatter(dra, ddec, s=5)
plt.plot(0, 0, "r+")
plt.plot((ra1-ra0) * 3.6e3, (dec1-dec0) * 3.6e3, "bx")
```
