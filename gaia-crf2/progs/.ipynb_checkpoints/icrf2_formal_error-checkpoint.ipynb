```python
from astropy.table import Table, join, Column
from astropy import units as u
import astropy.coordinates as coord
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# %config InlineBackend.figure_format = "svg"

# My modules
from my_progs.catalog.read_icrf import read_icrf2
```

```python
# Read ICRF2 data
icrf2 = read_icrf2()
icrf2
```

```python
# Defining sources
mask_def = (icrf2["type"] == "D")
icrf2_def = icrf2[mask_def]

# VCS sources
mask_vcs = (icrf2["type"] == "V")
icrf2_vcs = icrf2[mask_vcs]

# Non-VCS sources
mask_non = (icrf2["type"] == "N")
icrf2_non = icrf2[mask_non]

# Southern sources
mask_sou = (icrf2["dec"] < 0)
icrf2_sou = icrf2[mask_sou]

# Northern sources
mask_nor = (icrf2["dec"] >= 0)
icrf2_nor = icrf2[mask_nor]

# Non-VCS sources
mask_nov = (icrf2["type"] != "V")
icrf2_nov = icrf2[mask_nov]

# VCS + Non-VCS
mask_oth = (icrf2["type"] != "D")
icrf2_oth = icrf2[mask_oth]
```

```python
ra_def = coord.Angle(icrf2_def["ra"]).wrap_at(180*u.degree)
dec_def = coord.Angle(icrf2_def["dec"])

ra_oth = coord.Angle(icrf2_oth["ra"]).wrap_at(180*u.degree)
dec_oth = coord.Angle(icrf2_oth["dec"])

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra_def.radian, dec_def.radian, c="r", s=1, label="Defining")
ax.scatter(ra_oth.radian, dec_oth.radian, c="b", s=1, label="Other")
ax.set_xticklabels(["14h","16h","18h","20h","22h","0h","2h","4h","6h","8h","10h"])
ax.set_title("ICRF2 (3414 sources)")
ax.grid(True)
ax.legend(loc="upper right")

plt.savefig("../plots/icrf2_source_distribution.eps", dpi=100)
plt.savefig("../plots/icrf2_source_distribution.png")
```

```python
# The distribution of formal errors.
ra_all = coord.Angle(icrf2["ra"]).wrap_at(180*u.degree)
dec_all = coord.Angle(icrf2["dec"])

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
sc = ax.scatter(ra_all.radian, dec_all.radian, c=icrf2["ra_err"], 
                s=1, vmin=0, vmax=0.5, cmap="YlGnBu")
ax.set_xticklabels(["14h","16h","18h","20h","22h","0h","2h","4h","6h","8h","10h"])
plt.colorbar(sc, extend="max",
             label="$\sigma_{\\alpha^*}$ [mas]", fraction=0.02)
ax.set_title("ICRF2")
ax.grid(True)

plt.savefig("../plots/icrf2_ra_error.eps", dpi=100)
plt.savefig("../plots/icrf2_ra_error.png")
```

```python
# The distribution of formal errors.
ra_all = coord.Angle(icrf2["ra"]).wrap_at(180*u.degree)
dec_all = coord.Angle(icrf2["dec"])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="mollweide")
sc = ax.scatter(ra_all.radian, dec_all.radian, c=icrf2["dec_err"],
                s=1, vmin=0, vmax=0.5, cmap="YlGnBu")
ax.set_xticklabels(["14h", "16h", "18h", "20h", "22h",
                    "0h", "2h", "4h", "6h", "8h", "10h"])
plt.colorbar(sc, extend="max",
             label="$\sigma_{\delta}$ [mas]", fraction=0.02)
ax.set_title("ICRF2")
ax.grid(True)

plt.savefig("../plots/icrf2_dec_error.eps", dpi=100)
plt.savefig("../plots/icrf2_dec_error.png")
```

```python
# The distribution of formal errors.
ra_all = coord.Angle(icrf2["ra"]).wrap_at(180*u.degree)
dec_all = coord.Angle(icrf2["dec"])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="mollweide")
sc = ax.scatter(ra_all.radian, dec_all.radian, c=icrf2["pos_err"],
                s=1, vmin=0, vmax=0.5, cmap="YlGnBu")
ax.set_xticklabels(["14h", "16h", "18h", "20h", "22h",
                    "0h", "2h", "4h", "6h", "8h", "10h"])
plt.colorbar(sc, extend="max",
             label="$\sigma_{pos,max}$ [mas]", fraction=0.02)
ax.set_title("ICRF2")
ax.grid(True)

plt.savefig("../plots/icrf2_pos_error.eps", dpi=100)
plt.savefig("../plots/icrf2_pos_error.png")
```

```python
# The distribution of correlation between R.A. and Decl..
ra_all = coord.Angle(icrf2["ra"]).wrap_at(180*u.degree)
dec_all = coord.Angle(icrf2["dec"])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="mollweide")
sc = ax.scatter(ra_all.radian, dec_all.radian, c=icrf2["ra_dec_corr"],
                s=1, vmin=-1, vmax=1, cmap="bwr")
ax.set_xticklabels(["14h", "16h", "18h", "20h", "22h",
                    "0h", "2h", "4h", "6h", "8h", "10h"])
plt.colorbar(sc, ticks=np.arange(-1., 1.1, 0.5), format="$%+.1f$",
             label="Correlation", fraction=0.025)
ax.set_title("ICRF2")
ax.grid(True)

plt.savefig("../plots/icrf2_ra_dec_corr.eps", dpi=100)
plt.savefig("../plots/icrf2_ra_dec_corr.png")
```

```python
# Defining sources
mask_def = (icrf2["type"] == "D")
icrf2_def = icrf2[mask_def]

# VCS sources
mask_vcs = (icrf2["type"] == "V")
icrf2_vcs = icrf2[mask_vcs]

# Non-VCS sources
mask_non = (icrf2["type"] == "N")
icrf2_non = icrf2[mask_non]

# Southern sources
mask_sou = (icrf2["dec"] < 0)
icrf2_sou = icrf2[mask_sou]

# Northern sources
mask_nor = (icrf2["dec"] >= 0)
icrf2_nor = icrf2[mask_nor]

# Non-VCS sources
mask_nov = (icrf2["type"] != "V")
icrf2_nov = icrf2[mask_nov]



# Plot the median formal error
# Median formal error
med_ra = np.array([
    np.median(icrf2["ra_err"]),
    np.median(icrf2_def["ra_err"]),
    np.median(icrf2_vcs["ra_err"]),
    np.median(icrf2_non["ra_err"]),
    np.median(icrf2_nor["ra_err"]),
    np.median(icrf2_sou["ra_err"]),
    np.median(icrf2_nov["ra_err"])
])

med_dec = np.array([
    np.median(icrf2["dec_err"]),
    np.median(icrf2_def["dec_err"]),
    np.median(icrf2_vcs["dec_err"]),
    np.median(icrf2_non["dec_err"]),
    np.median(icrf2_nor["dec_err"]),
    np.median(icrf2_sou["dec_err"]),
    np.median(icrf2_nov["dec_err"])
])

med_pos = np.array([
    np.median(icrf2["pos_err"]),
    np.median(icrf2_def["pos_err"]),
    np.median(icrf2_vcs["pos_err"]),
    np.median(icrf2_non["pos_err"]),
    np.median(icrf2_nor["pos_err"]),
    np.median(icrf2_sou["pos_err"]),
    np.median(icrf2_nov["pos_err"])
])

names = ["All", "Defining", "VCS", "Non-VCS", "North", "South", "Def+Non"]

pos = np.arange(len(names)) + 1
width = 0.8 / (2 * len(names) - 1)

fig, ax = plt.subplots()
ax.bar(pos - width, med_ra, width=width, label="$\mathrm{\sigma_{\\alpha^*}}$")
ax.bar(pos, med_dec, width=width, label="$\mathrm{\sigma_{\\delta}}$")
ax.bar(pos + width, med_pos, width=width, label="$\mathrm{\sigma_{pos,max}}$")

ax.set_xticks(pos)
ax.set_xticklabels(names)

ax.set_ylabel("Median formal error [mas]")

ax.set_title("ICRF2 catalog")

ax.legend()
ax.grid()

plt.savefig("../plots/icrf2_median_error.eps")
```

```python
print("The median formal error in the ICRF2 (mas)")
print("          RA*    DEC    EEMA")
print("All       {:.3f}  {:.3f}  {:.3f}".format(
    np.median(icrf2["ra_err"]),
    np.median(icrf2["dec_err"]),
    np.median(icrf2["pos_err"])))
print("Defining  {:.3f}  {:.3f}  {:.3f}".format(
    np.median(icrf2_def["ra_err"]),
    np.median(icrf2_def["dec_err"]),
    np.median(icrf2_def["pos_err"])))
print("VCS       {:.3f}  {:.3f}  {:.3f}".format(
    np.median(icrf2_vcs["ra_err"]),
    np.median(icrf2_vcs["dec_err"]),
    np.median(icrf2_vcs["pos_err"])))
print("Non-VCS   {:.3f}  {:.3f}  {:.3f}".format(
    np.median(icrf2_non["ra_err"]),
    np.median(icrf2_non["dec_err"]),
    np.median(icrf2_non["pos_err"])))
```
