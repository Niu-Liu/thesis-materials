```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# File name: catalog_error_comparison.py
"""
Created on Sun May 27 15:28:04 2018

@author: Neo(liuniu@smail.nju.edu.cn)

Plot the smoothed formal error as a function of declination of various catalogs.

"""

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from matplotlib import pyplot as plt
import numpy as np
# My modules
from read_icrf1 import read_icrf1_pos
from read_icrf2 import read_icrf2
from read_sou import read_cat
from read_GaiaDR1 import read_gaiadr1
from read_GaiaDR2 import read_gaiadr2_iers_position
from cross_match import (list_crossmatch, pos_max_calc,
                         overall_err_calc, postional_difference_calc)
from declination_smoothed import dec_smoothed_pos_max
```

```python
# ICRF1 catalog
[icrf_name_i1, iers_name_i1,
 ra_i1, dec_i1, ra_error_i1, dec_error_i1, ra_dec_corr_i1, _] = read_icrf1_pos("../data/rsc95r01.dat")

# ellipse semi-major axis
sig_pos_max_i1 = pos_max_calc(ra_error_i1, dec_error_i1, ra_dec_corr_i1)

# overall formal uncertainty
overall_err_i1 = overall_err_calc(ra_error_i1, dec_error_i1, ra_dec_corr_i1)
```

```python
# ICRF2 catalog
[icrf_name_i2, ivs_name_i2, iers_name_i2,
 ra_i2, dec_i2, ra_error_i2, dec_error_i2, ra_dec_corr_i2, _, _] = read_icrf2("../data/icrf2.dat")

# ellipse semi-major axis
sig_pos_max_i2 = pos_max_calc(ra_error_i2, dec_error_i2, ra_dec_corr_i2)

# overall formal uncertainty
overall_err_i2 = overall_err_calc(ra_error_i2, dec_error_i2, ra_dec_corr_i2)
```

```python
# VLBI solution OPAA
vlbi_cat_v1 = "../data/opa-sx-180425-noGA.cat"
[ivs_name_v1, iers_name_v1, ra_v1, dec_v1, ra_error_v1, dec_error_v1,
 ra_dec_corr_v1, num_ses_v1, num_obs_v1] = read_cat(vlbi_cat_v1)

# ellipse semi-major axis
sig_pos_max_v1 = pos_max_calc(ra_error_v1, dec_error_v1, ra_dec_corr_v1)

# overall formal uncertainty
overall_err_v1 = overall_err_calc(ra_error_v1, dec_error_v1, ra_dec_corr_v1)
```

```python
# VLBI solution OPAB
vlbi_cat_v2 = "../data/opa-sx-180425-GA00.cat"
[ivs_name_v2, iers_name_v2, ra_v2, dec_v2, ra_error_v2, dec_error_v2,
 ra_dec_corr_v2, num_ses_v2, num_obs_v2] = read_cat(vlbi_cat_v2)

# ellipse semi-major axis
sig_pos_max_v2 = pos_max_calc(ra_error_v2, dec_error_v2, ra_dec_corr_v2)

# overall formal uncertainty
overall_err_v2 = overall_err_calc(ra_error_v2, dec_error_v2, ra_dec_corr_v2)
```

```python
# VLBI solution OPAC
vlbi_cat_v3 = "../data/opa-sx-180425-GA15.cat"
[ivs_name_v3, iers_name_v3, ra_v3, dec_v3, ra_error_v3, dec_error_v3,
 ra_dec_corr_v3, num_ses_v3, num_obs_v3] = read_cat(vlbi_cat_v3)

# ellipse semi-major axis
sig_pos_max_v3 = pos_max_calc(ra_error_v3, dec_error_v3, ra_dec_corr_v3)

# overall formal uncertainty
overall_err_v3 = overall_err_calc(ra_error_v3, dec_error_v3, ra_dec_corr_v3)
```

```python
# Load Gaia DR1 data
[icrf_name_g1, ra_g1, dec_g1, 
 ra_error_g1, dec_error_g1, ra_dec_corr_g1, _] = read_gaiadr1("../data/qso.dat")

# ellipse semi-major axis
sig_pos_max_g1 = pos_max_calc(ra_error_g1, dec_error_g1, ra_dec_corr_g1)

# overall formal uncertainty
overall_err_g1 = overall_err_calc(ra_error_g1, dec_error_g1, ra_dec_corr_g1)
```

```python
# Load Gaia DR2 data
gaia_cat2 = "../data/gaiadr2_iers.fits"
[iers_name_g2, ra_g2, ra_error_g2,
 dec_g2, dec_error_g2, ra_dec_corr_g2] = read_gaiadr2_iers_position(gaia_cat2)

# ellipse semi-major axis
sig_pos_max_g2 = pos_max_calc(ra_error_g2, dec_error_g2, ra_dec_corr_g2)

# overall formal uncertainty
overall_err_g2 = overall_err_calc(ra_error_g2, dec_error_g2, ra_dec_corr_g2)
```

```python
# Plot for median error
fig, ax = plt.subplots()

ax.plot(dec_g1, sig_pos_max_g1, ".", ms=1, label="Gaia DR1")
ax.plot(dec_g2, sig_pos_max_g2, ".", ms=1, label="Gaia DR2")
ax.plot(dec_i1, sig_pos_max_i1, ".", ms=1, label="ICRF1")
ax.plot(dec_i2, sig_pos_max_i2, ".", ms=1, label="ICRF2")
ax.plot(dec_v1, sig_pos_max_v1, ".", ms=1, label="OPAA")
# ax.plot(dec_v2, sig_pos_max_v2, ".", ms=1, label="OPAB")
# ax.plot(dec_v3, sig_pos_max_v3, ".", ms=1, label="OPAC")

ax.set_xlim([-90, 90])
ax.set_xticks(np.arange(-90, 91, 30))
ax.set_xlabel("Declination (degree)")
ax.set_yscale("log")
ax.set_ylabel('Error (mas)')
ax.grid()  
ax.legend(fontsize="small")
```

```python
# Calculate the smoothed standard errors

bin_size = 60

# ICRF1
med_dec_i1, med_poserr_i1 = dec_smoothed_pos_max(dec_i1, sig_pos_max_i1, bin_size)

# ICRF2
med_dec_i2, med_poserr_i2 = dec_smoothed_pos_max(dec_i2, sig_pos_max_i2, bin_size)

# OPA-A
med_dec_v1, med_poserr_v1 = dec_smoothed_pos_max(dec_v1, sig_pos_max_v1, bin_size)

# OPA-B
med_dec_v2, med_poserr_v2 = dec_smoothed_pos_max(dec_v2, sig_pos_max_v2, bin_size)

# OPA-C
med_dec_v3, med_poserr_v3 = dec_smoothed_pos_max(dec_v3, sig_pos_max_v3, bin_size)

# Gaia DR1
med_dec_g1, med_poserr_g1 = dec_smoothed_pos_max(dec_g1, sig_pos_max_g1, bin_size)

# Gaia DR2
med_dec_g2, med_poserr_g2 = dec_smoothed_pos_max(dec_g2, sig_pos_max_g2, bin_size)
```

```python
# Plot for median error
fig, ax = plt.subplots()

ax.plot(med_dec_g1, med_poserr_g1, ".", ms=2, label="Gaia DR1")
ax.plot(med_dec_g2, med_poserr_g2, "k.", ms=2, label="Gaia DR2")
ax.plot(med_dec_i1, med_poserr_i1, "b.", ms=2, label="ICRF1")
ax.plot(med_dec_i2, med_poserr_i2, "y.", ms=2, label="ICRF2")
ax.plot(med_dec_v1, med_poserr_v1, ".", ms=2, label="OPAA")
# ax.plot(med_dec_v2, med_poserr_v2, ".", ms=2, label="OPAB")
# ax.plot(med_dec_v3, med_poserr_v3, ".", ms=2, label="OPAC")

ax.set_xlim([-90, 90])
ax.set_xticks(np.arange(-90, 91, 30))
ax.set_xlabel("Declination (degree)")
ax.set_yscale("log")
ax.set_ylabel('Error (mas)')
ax.grid()  
ax.legend(fontsize="small")
```
