```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# File name: VLBI_internal_error4.py
"""
Created on Fri Jun 22 17:26:33 2018

@author: Neo(liuniu@smail.nju.edu.cn)


Use Sebastien's method to estimate noise floor and scale factor

"""

import matplotlib.pyplot as plt
import numpy as np
from numpy import cos, deg2rad, sqrt
from scipy.optimize import curve_fit
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
# My modules
from read_icrf2 import read_icrf2
from error_inflation_progs import (error_inflation, offset_wrms_errbinned, wrms_calc)
```

```python
def offset_wrms_errbinned(x, x_err, interv=0.04, max_err=1):
    """Calculate the scatter of positional offset binned by formal error

    Parameters
    ----------
    x : array_like of float
        data
    x_err : array_like of float
        formal error of x
    interv : float
        width of interval in mas
    max_err : float
        maximum of x_err in mas

    Returns
    ----------
    x_wrms : float of array
        wrms of x binned by interval of x_err
    err_med : float of array
        dedian of interval of x_err
    err_mid : float of array
        mid-point of interval of x_err
    """

    # interv = 40  # uas
    # max_err = 1000  # uas
    num_interv = int(max_err / interv)
    x_wrms = np.zeros(num_interv)
    err_mid = np.ones(num_interv)
    err_med = np.ones(num_interv)

    for i in range(num_interv):
        interv_b = 0.001 + i * interv
        interv_e = interv_b + interv
        con = (x_err > interv_b) & (x_err <= interv_e)
        xs = x[con]
        x_errs = x_err[con]

#         x_wrms[i] = wrms_calc(xs, x_errs)
        x_wrms[i] = np.std(xs)
        err_mid[i] = (interv_b + interv_e) / 2.
        err_med[i] = np.median(x_errs)
        
        print("(%.3f, %.3f] mas:  %4d sources"%(interv_b, interv_e, xs.size))

    return x_wrms, err_med, err_mid
```

```python
# Load data
datfile = ("../data/odd-even.cat_diff")
soulist = np.genfromtxt(datfile, usecols=(0,), dtype=str)
RA, Dec, dRAc, dDec, dRAc_err, dDec_err = np.genfromtxt(
    datfile, usecols=range(1, 7), unpack=True)
num_ses_mean, num_obs_mean = np.genfromtxt(
    datfile, dtype=int, usecols=range(7, 9), unpack=True)

# scale the position offset by their uncertainty
nor_dRAc = dRAc / dRAc_err
nor_dDec = dDec / dDec_err
```

```python
# Calculate the offset and normalized seperation
rho = sqrt(dRAc**2 + dDec**2)
nor_rho = sqrt(nor_dRAc**2 + nor_dDec**2)
rho_err = sqrt(dRAc_err**2 + dDec_err**2)

# Calculate the positional offset scatter
# interv = 0.02
dRAc_wrms, dRAc_err_med, dRAc_err_mid = offset_wrms_errbinned(
    dRAc, dRAc_err, interv=0.01, max_err=1)
dDec_wrms, dDec_err_med, dDec_err_mid = offset_wrms_errbinned(
    dDec, dDec_err, interv=0.01, max_err=1)
```

```python
# Use data in [0.001, 0.1] mas
con = (dRAc_err_med <= 0.1) & (dDec_err_med <= 0.07)

# Edit data
dRAc_err_med_s = dRAc_err_med[con]
dRAc_wrms_s = dRAc_wrms[con]
dDec_err_med_s = dDec_err_med[con]
dDec_wrms_s = dDec_wrms[con]
# Manually remove 8th data point
# dDec_err_med_s[4] = dDec_wrms_s[4]
# dDec_err_med_s[7] = dDec_wrms_s[7]

# estimate the noise floor and scale factor
init_vals = [1.0, 0.04]

# For R.A. component
best_vals_RA, covar_RA = curve_fit(
    error_inflation, dRAc_err_med_s, dRAc_wrms_s, p0=init_vals)
sf_RA, nf_RA = best_vals_RA
sf_RA_err, nf_RA_err = np.sqrt(np.diag(covar_RA))
dRAc_err_inf = error_inflation(dRAc_err_med, sf_RA, nf_RA)

print("For R.A. component:\n"
      "scaling factor: %.3f +/- %.3f\n"
      "noise floor: %.3f +/- %.3f mas\n" %
      (sf_RA, sf_RA_err, nf_RA, nf_RA_err))

# For Dec. component
best_vals_Dec, covar_Dec = curve_fit(
    error_inflation, dDec_err_med_s, dDec_wrms_s, p0=init_vals)
sf_Dec, nf_Dec = best_vals_Dec
sf_Dec_err, nf_Dec_err = np.sqrt(np.diag(covar_Dec))
dDec_err_inf = error_inflation(dDec_err_med, sf_Dec, nf_Dec)

print("For Dec. component:\n"
      "scaling factor: %.3f +/- %.3f\n"
      "noise floor: %.3f +/- %.3f mas\n" %
      (sf_Dec, sf_Dec_err, nf_Dec, nf_Dec_err))
```

```python
# Plot the median error, wrms difference, and inflated errors

print("Scale factors of RA and Dec. are %.3f and %.3f." % (sf_RA, sf_Dec))
print("Noise floors of RA and Dec. are %.3f and %.3f mas." % (nf_RA, nf_Dec))

# wrms difference VS median formal uncertainty (all sources)
fig, (ax0, ax1) = plt.subplots(figsize=(10, 4), ncols=2, sharex=True)

ax0.plot(dRAc_err_med, dRAc_wrms, "bo", markersize=2, label="WRMS")
ax0.plot(dRAc_err_med, dRAc_wrms, "g", lw=1)
ax0.plot(dRAc_err_med, dRAc_err_inf, "r-", markersize=2, label="Inflated error")
ax1.plot(dDec_err_med, dDec_wrms, "bo", markersize=2, label="WRMS")
ax1.plot(dDec_err_med, dDec_wrms, "g", lw=1)
ax1.plot(dDec_err_med, dDec_err_inf, "r-", markersize=2, label="Inflated error")

N = np.arange(0.001, 1, 0.01)
ax0.plot(N, N, "k", lw=0.5)
ax1.plot(N, N, "k", lw=0.5)

ax0.set_xscale("log")
ax0.set_yscale("log")
ax0.set_xlabel("Formal error in $\\alpha\\,\\cos\\delta$ (mas)")
ax0.set_ylabel("WRMS in $\\alpha\\,\\cos\\delta$ (mas)")
ax0.set_xlim([0.001, 1])
ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax0.grid(True)
ax0.legend()

ax1.set_xscale("log")
ax1.set_yscale("log")
ax1.set_xlabel("Formal error in $\\delta$ (mas)")
ax1.set_ylabel("WRMS in $\\delta$(mas)")
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')
ax1.grid(True)
ax1.legend()


plt.tight_layout()
# plt.subplots_adjust(hspace=0.1)

# plt.savefig("/Users/Neo/Astronomy/Works/201711_GDR2_ICRF3/plots/wrms_median_session_all.eps")
```

```python
# Remove outliers
# 1) offset < 1 mas and offset to its uncertainty < 5 (used in Sebstien's paper)
# con = (rho <= 0.5) & (rho/rho_err <= 5)
con = (rho <= 1) & (nor_rho <= 5)
print("There are %d sources with offset < 1. mas and offset to its uncertainty < 5." %
      soulist[con].size)

# edit data
dRAc_s = dRAc[con]
dRAc_err_s = dRAc_err[con]
dDec_s = dDec[con]
dDec_err_s = dDec_err[con]


# Calculate the positional offset scatter
# interv = 0.02
dRAc_wrms_s, dRAc_err_med_s, dRAc_err_mid_s = offset_wrms_errbinned(
    dRAc_s, dRAc_err_s, interv=0.03, max_err=1)
dDec_wrms_s, dDec_err_med_s, dDec_err_mid_s = offset_wrms_errbinned(
    dDec_s, dDec_err_s, interv=0.03, max_err=1)

# # Use data in [0.001, 0.2] mas
con = (dRAc_err_med_s <= 0.2) & (dDec_err_med_s <= 0.2)

# Edit data
dRAc_err_med_s1 = dRAc_err_med_s[con]
dRAc_wrms_s1 = dRAc_wrms_s[con]
dDec_err_med_s1 = dDec_err_med_s[con]
dDec_wrms_s1 = dDec_wrms_s[con]

# estimate the noise floor and scale factor
init_vals = [1.0, 0.04]
# For R.A. component
best_vals_RA, covar_RA = curve_fit(
    error_inflation, dRAc_err_med_s1, dRAc_wrms_s1, p0=init_vals)
sf_RA, nf_RA = best_vals_RA
sf_RA_err, nf_RA_err = np.sqrt(np.diag(covar_RA))
corr_RA = covar_RA[0, 1] / sf_RA_err / nf_RA_err
dRAc_err_inf_s = error_inflation(dRAc_err_med_s, sf_RA, nf_RA)

print("For R.A. component:\n"
      "scaling factor: %.2f +/- %.2f\n"
      "noise floor: %.3f +/- %.3f mas\n"
      "correlation coefficent: %+.2f" %
      (sf_RA, sf_RA_err, nf_RA, nf_RA_err, corr_RA))

# For Dec. component
best_vals_Dec, covar_Dec = curve_fit(
    error_inflation, dDec_err_med_s1, dDec_wrms_s1, p0=init_vals)
sf_Dec, nf_Dec = best_vals_Dec
sf_Dec_err, nf_Dec_err = np.sqrt(np.diag(covar_Dec))
corr_Dec = covar_Dec[0, 1] / sf_Dec_err / nf_Dec_err
dDec_err_inf_s = error_inflation(dDec_err_med_s, sf_Dec, nf_Dec)

print("For Dec. component:\n"
      "scaling factor: %.2f +/- %.2f\n"
      "noise floor: %.3f +/- %.3f mas\n"
      "correlation coefficent: %+.2f" %
      (sf_Dec, sf_Dec_err, nf_Dec, nf_Dec_err, corr_Dec))
```

```python
# Plot the median error, wrms difference, and inflated errors

print("Scale factors of RA and Dec. are %.3f and %.3f." % (sf_RA, sf_Dec))
print("Noise floors of RA and Dec. are %.3f and %.3f mas." % (nf_RA, nf_Dec))

# wrms difference VS median formal uncertainty (all sources)
# fig, (ax0, ax1) = plt.subplots(figsize=(10, 4), ncols=2, sharex=True)
fig, (ax0, ax1) = plt.subplots(figsize=(4, 6), nrows=2, sharex=True)

ax0.plot(dRAc_err_med_s, dRAc_wrms_s, "bo-", markersize=5, label="Rms scatter", lw=0.5)
# ax0.plot(dRAc_err_med_s, dRAc_wrms_s, "b-", lw=1)
ax0.plot(dRAc_err_med_s, dRAc_err_inf_s, "r--", markersize=2, label="Inflated error")
ax1.plot(dDec_err_med_s, dDec_wrms_s, "bo-", markersize=5, label="Rms scatter", lw=0.5)
# ax1.plot(dDec_err_med_s, dDec_wrms_s, "g", lw=1)
ax1.plot(dDec_err_med_s, dDec_err_inf_s, "r--", markersize=2, label="Inflated error")

N = np.arange(0.001, 1, 0.01)
ax0.plot(N, N, "k", lw=0.5)
ax1.plot(N, N, "k", lw=0.5)

ax0.set_xscale("log")
ax0.set_yscale("log")
ax0.set_xlabel("Median formal error in $\\alpha\\,\\cos\\delta$ (mas)")
ax0.set_ylabel("Scatter of difference in $\\alpha\\,\\cos\\delta$ (mas)")
ax0.set_xlim([0.01, 1])
ax0.set_ylim([0.01, 2])
ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax0.grid(True)
ax0.legend()

ax1.set_xscale("log")
ax1.set_yscale("log")
ax1.set_ylim([0.01, 2])
ax1.set_xlabel("Median formal error in $\\delta$ (mas)")
ax1.set_ylabel("Scatter of difference in $\\delta$(mas)")
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')
ax1.grid(True)
# ax1.legend()

plt.tight_layout()
plt.savefig("../plots/error-SBL.eps")
```
