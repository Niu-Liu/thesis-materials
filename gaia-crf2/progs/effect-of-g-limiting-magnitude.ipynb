```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = "svg"
```

The number of common sources as a function of limiting magnitude.

```python
gmag3sx, N0_3sx, N1_3sx, X0_3sx, rate3sx = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf3sx.log", unpack=True)
gmag3k, N0_3k, N1_3k, X0_3k, rate3k = np.genfromtxt(
    "../logs/g-limiting-mag-stat-icrf3k.log", unpack=True)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(gmag3sx, N0_3sx, "r-^", label="ICRF3 S/X (Common)", ms=3, lw=0.5)
ax.plot(gmag3sx, N1_3sx, "r-o", label="ICRF3 S/X (Used)", ms=3, lw=0.5)

ax.plot(gmag3k, N0_3k, "b-^", label="ICRF3 K (Common)", ms=3, lw=0.5)
ax.plot(gmag3k, N1_3k, "b-o", label="ICRF3 K (Used)", ms=3, lw=0.5)

ax.axis([17.4, 21.1, 0, 3000])
ax.set_xlabel("$G_{\\rm max}$ (mag)")
ax.set_ylabel("No. sources")

ax.legend(fontsize="x-small")

plt.tight_layout()
plt.savefig("../plots/gdr2-nosou-vs-G.eps", hbox="tight")
```

VSH parameters vs. $G$-magnitude

1. Rotation

2. Dipole

3. Magnetic parameters of second-orders

```python
par3sx = np.genfromtxt("../logs/g-limiting-mag-vsh02-icrf3sx.log")
par3k = np.genfromtxt("../logs/g-limiting-mag-vsh02-icrf3k.log")
```

```python
# Rotation
fig, (ax0, ax1, ax2) = plt.subplots(figsize=(6, 4), nrows=3, sharex=True)

# R1
ax0.errorbar(par3sx[:, 0], par3sx[:, 4], yerr=par3sx[:, 20],
            fmt="r-^", label="ICRF3 S/X", capsize=2, lw=0.5, elinewidth=0.5, ms=3)
ax0.errorbar(par3k[:, 0], par3k[:, 4], yerr=par3k[:, 20],
            fmt="b-o", label="ICRF3 K", capsize=2, lw=0.5, elinewidth=0.5, ms=3)

# R2
ax1.errorbar(par3sx[:, 0], par3sx[:, 5], yerr=par3sx[:, 21],
            fmt="r-^", label="ICRF3 S/X", capsize=2, lw=0.5, elinewidth=0.5, ms=3)
ax1.errorbar(par3k[:, 0], par3k[:, 5], yerr=par3k[:, 21],
            fmt="b-o", label="ICRF3 K", capsize=2, lw=0.5, elinewidth=0.5, ms=3)

# R3
ax2.errorbar(par3sx[:, 0], par3sx[:, 6], yerr=par3sx[:, 22],
            fmt="r-^", label="ICRF3 SX", capsize=2, lw=0.5, elinewidth=0.5, ms=3)
ax2.errorbar(par3k[:, 0], par3k[:, 6], yerr=par3k[:, 22],
            fmt="b-o", label="ICRF3 K", capsize=2, lw=0.5, elinewidth=0.5, ms=3)

# ax0.axis([17.4, 21.1, -10, 80])
ax0.set_xlim([17.4, 21.1])
ax2.set_xlabel("$G_{\\rm max}$ (mag)")
ax0.set_ylabel("$R_1$ ($\mathrm{\mu}$as)")
ax1.set_ylabel("$R_2$ ($\mathrm{\mu}$as)")
ax2.set_ylabel("$R_3$ ($\mathrm{\mu}$as)")

ax2.yaxis.set_ticks_position("both")

ax0.legend(loc=2, bbox_to_anchor=(1.01, 1.0), borderaxespad=0., fontsize="x-small")

plt.tight_layout()
plt.subplots_adjust(hspace=0)

plt.savefig("../plots/gdr2-rotation-vs-G.eps", hbox="tight")
```
