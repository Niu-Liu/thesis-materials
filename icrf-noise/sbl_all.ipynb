```python
def formal_error_binner(x, x_err, binsize=50, step=50):
    """Calculate the scatter of positional offset binned by formal error

    Parameters
    ----------
    x : array_like of float
        data
    x_err : array_like of float
        formal error of x
    binsize : int
        number of points in every bin
    max_err : float
        maximum of x_err in mas

    Returns
    ----------
    x_wrms : float of array
        wrms of x binned by interval of x_err
    err_med : float of array
        median of interval of x_err
    """

#     if x.size % binsize:
#         num_bin = x.size // binsize + 1
#     else:
#         num_bin = x.size // binsize
        
    if (x.size - binsize) % step:
        num_bin = (x.size - binsize) // step + 1
    else:
        num_bin = (x.size - binsize) // step

    x_wrms = np.zeros(num_bin)
    err_med = np.ones(num_bin)

    # Sorted x by x_err
    indx = np.argsort(x_err)
    x_sort = np.take(x, indx)
    x_err_sort = np.take(x_err, indx)

    for i in range(num_bin-1):
        xs = x_sort[i * step: i * step + binsize]
        x_errs = x_err_sort[i * step: i * step + binsize]

        x_wrms[i] = wrms_calc(xs, x_errs)
        err_med[i] = np.median(x_errs)

    x_wrms[-1] = wrms_calc(x_sort[(num_bin-1) * step:],
                           x_err_sort[(num_bin-1) * step:])
    err_med[-1] = np.median(x_err_sort[(num_bin-1) * step:])

    return x_wrms, err_med
```

```python
# Load data
datfile = ("../logs/pos_diff_cln01.dat")
soulist = np.genfromtxt(datfile, usecols=(0,), dtype=str)
RA, Dec, dRAc, dDec, dRAc_err, dDec_err = np.genfromtxt(
    datfile, usecols=range(1, 7), unpack=True)
num_ses_mean, num_obs_mean = np.genfromtxt(
    datfile, dtype=int, usecols=range(10, 12), unpack=True)

# Calculate the positional offset scatter
dRAc_wrms, dRAc_err_med = formal_error_binner(
    dRAc, dRAc_err, binsize=50, step=50)
dDec_wrms, dDec_err_med = formal_error_binner(
    dDec, dDec_err, binsize=50, step=50)
```

```python
# Use data in [0.001, 0.1] mas
err_max = 0.75
con = (dRAc_err_med <= err_max) & (dDec_err_med <= err_max)

# Edit data
dRAc_err_med_s = dRAc_err_med[con]
dRAc_wrms_s = dRAc_wrms[con]
dDec_err_med_s = dDec_err_med[con]
dDec_wrms_s = dDec_wrms[con]

# estimate the noise floor and scale factor
init_vals = [1.0, 0.04]

# For R.A. component
best_vals_RA, covar_RA = curve_fit(
    error_inflation, dRAc_err_med_s, dRAc_wrms_s, p0=init_vals)
sf_RA, nf_RA = best_vals_RA
sf_RA_err, nf_RA_err = np.sqrt(np.diag(covar_RA))
nf_RA = nf_RA / sqrt(2)
nf_RA_err = nf_RA_err / sqrt(2)

print("For R.A. component:\n"
      "scaling factor: %.3f +/- %.3f\n"
      "noise floor: %.3f +/- %.3f mas\n" %
      (sf_RA, sf_RA_err, nf_RA, nf_RA_err))

# For Dec. component
best_vals_Dec, covar_Dec = curve_fit(
    error_inflation, dDec_err_med_s, dDec_wrms_s, p0=init_vals)
sf_Dec, nf_Dec = best_vals_Dec
sf_Dec_err, nf_Dec_err = np.sqrt(np.diag(covar_Dec))
nf_Dec = nf_Dec / sqrt(2)
nf_Dec_err = nf_Dec_err / sqrt(2)

print("For Dec. component:\n"
      "scaling factor: %.3f +/- %.3f\n"
      "noise floor: %.3f +/- %.3f mas\n" %
      (sf_Dec, sf_Dec_err, nf_Dec, nf_Dec_err))
```

```python
# Plot the median error, wrms difference, and inflated errors

print("Scale factors of RA and Dec. are %.2f and %.2f." % (sf_RA, sf_Dec))
print("Noise floors of RA and Dec. are %.3f and %.3f mas." % (nf_RA, nf_Dec))

N = np.arange(1, 10.e3, 1)
dRAc_err_inf = error_inflation(N, sf_RA, nf_RA * 1.e3)
dDec_err_inf = error_inflation(N, sf_Dec, nf_Dec * 1.e3)

# wrms difference VS median formal uncertainty (all sources)
# fig, (ax0, ax1) = plt.subplots(figsize=(10, 4), ncols=2, sharex=True)
fig, (ax0, ax1) = plt.subplots(figsize=(5, 6), nrows=2)

ax0.plot(dRAc_err_med * 1.e3, dRAc_wrms * 1.e3, "bo-", 
         markersize=3, label="WRMS scatter", lw=0.5)
ax0.plot(N, dRAc_err_inf, "r--", markersize=2, label="Inflated error (SBL)")
ax1.plot(dDec_err_med * 1.e3, dDec_wrms * 1.e3, "bo-", 
         markersize=3, label="WRMS scatter", lw=0.5)
ax1.plot(N, dDec_err_inf, "r--", markersize=2, label="Inflated error (SBL)")

ax0.plot(N, N, "k", lw=0.5)
ax1.plot(N, N, "k", lw=0.5)

ax0.set_xscale("log")
ax0.set_yscale("log")
ax0.set_xlabel("Median formal error in $\\alpha\\,\\cos\\delta$ [$\mu$as]")
ax0.set_ylabel("WRMS scatter in $\\alpha\\,\\cos\\delta$ [$\mu$as]")
# ax0.set_xlim([0.001, 1])
# ax0.set_ylim([0.001, 1])
ax0.set_xlim([1, 1.e3])
ax0.set_ylim([1, 1.e3])
ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax0.grid(True)
ax0.legend(fontsize="small")

ax1.set_xlim([1, 1.e3])
ax1.set_xscale("log")
ax1.set_yscale("log")
# ax1.set_ylim([0.001, 1])
ax1.set_ylim([1, 1.e3])
ax1.set_xlabel("Median formal error in $\\delta$ [$\mu$as]")
ax1.set_ylabel("WRMS scatter in $\\delta$ [$\mu$as]")
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')
ax1.grid(True)
ax1.legend(fontsize="small")

plt.tight_layout()
plt.subplots_adjust(hspace=0.2)
# plt.savefig("../plots/SBL.eps")
# plt.savefig("/Users/Neo/Desktop/noise-sbl.png")
```
