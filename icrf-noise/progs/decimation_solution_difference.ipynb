In this notebook, I do some statistical plots on the radio source position difference for the clean sample.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# File name: decimation_solution_difference.py
"""
Created on Thu Jun 21 09:30:04 2018

@author: Neo(liuniu@smail.nju.edu.cn)
"""

import matplotlib.pyplot as plt
import numpy as np
from numpy import cos, deg2rad, sqrt
from scipy.optimize import curve_fit
%matplotlib inline
%config InlineBackend.figure_format = "svg"
# My module
from my_progs.catalog.read_icrf import read_icrf2
```

First, load the Table of position offsets for the clean sample 

```python
# Load data
datfile = "../logs/pos_diff_cln.dat"
soulist = np.genfromtxt(datfile, usecols=(0,), dtype=str)
RA, Dec, dRAc, dDec, dRAc_err, dDec_err, eema, ang_sep, X = np.genfromtxt(
    datfile, usecols=range(1, 10), unpack=True)
num_ses_mean, num_obs_mean = np.genfromtxt(
    datfile, dtype=int, usecols=range(10, 12), unpack=True)

# scale the position offset by their uncertainty
nor_dRAc = dRAc / dRAc_err
nor_dDec = dDec / dDec_err
```

```python
# Find VCS sources

# ICRF2 data
_, ivsn, _, _, _, _, _, _, _, Flag = read_icrf2("../data/icrf2.dat")

# VCS source
con = (Flag == "V")
icrf2_vcs = ivsn[con]
    
# Find defining sources
vcssou = [sou for sou in soulist if sou in icrf2_vcs]
        
vcssou = np.asarray(vcssou)

print("There are %d common sources, %d of which are VCS sources." % (soulist.size, vcssou.size))
```

```python
# VCS source
con = (Flag == "V")
icrf2_vcs = ivsn[con]
    
# Find defining sources
sou_flg = []
for sou in soulist:
    if sou in icrf2_vcs:
#         sou_flg.append(False)
        sou_flg.append(True)
    else:
#         sou_flg.append(True)
        sou_flg.append(False)
            
sou_flg = np.asarray(sou_flg)

# nor_dRAc_novcs = nor_dRAc[sou_flg]
# nor_dDec_novcs = nor_dDec[sou_flg]
nor_dRAc_vcs = nor_dRAc[sou_flg]
nor_dDec_vcs = nor_dDec[sou_flg]

# print("Non-VCS Source number: ", nor_dRAc_novcs.size)
print("VCS Source number: ", nor_dRAc_vcs.size)
```

```python
# Plot for the difference
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4.2))

# bins = np.arange(-5, 5, 0.2)
bins = np.arange(-1., 1., 0.05)

ax0.hist(dRAc, bins, histtype='bar', rwidth=0.8, label="All")
ax0.hist(dRAc[sou_flg], bins, histtype='bar', rwidth=0.8, label="VCS")
ax0.set_ylabel("Number of Sources")
# ax0.set_ylim([0, 700])
ax0.set_ylim([0, 250])
ax0.set_xlabel("$\Delta \\alpha\cos\delta$ [mas]")
ax0.legend()

ax1.hist(dDec, bins, histtype='bar', rwidth=0.8, label="All")
ax1.hist(dDec[sou_flg], bins, histtype='bar', rwidth=0.8, label="VCS")
ax1.set_ylabel("Number of Sources")
# ax1.set_ylim([0, 700])
ax1.set_ylim([0, 250])
ax1.set_xlabel("$\Delta \delta$ [mas]")
# ax1.legend()

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
minorLocator = MultipleLocator(1)
ax0.xaxis.set_minor_locator(minorLocator)
ax1.xaxis.set_minor_locator(minorLocator)

ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')

fig.tight_layout()
```

```python
# Plot for the normalized difference
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4.2))
# fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(4, 6))

# bins = np.arange(-10., 10., 0.5)
bins = np.arange(-5., 5., 0.5)

ax0.hist(nor_dRAc, bins, histtype='bar', rwidth=0.8, label="All")
ax0.hist(nor_dRAc[sou_flg], bins, histtype='bar', rwidth=0.8, label="VCS")
ax0.set_xlabel("$\\Delta\\alpha\cos\delta/\sigma_{\\Delta\\alpha\cos\delta}$")
ax0.set_ylabel("Number of Sources")
ax0.set_ylim([0, 450])
ax0.legend()

ax1.hist(nor_dDec, bins, histtype='bar', rwidth=0.8, label="All")
ax1.hist(nor_dDec[sou_flg], bins, histtype='bar', rwidth=0.8, label="VCS")

ax1.set_ylim([0, 450])
ax1.set_xlabel("$\Delta\delta/\sigma_{\Delta\delta}$")
ax1.set_ylabel("Number of Sources")
# ax1.legend()

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
minorLocator = MultipleLocator(1)
ax0.xaxis.set_minor_locator(minorLocator)
ax1.xaxis.set_minor_locator(minorLocator)

minorLocator = MultipleLocator(10)
ax0.yaxis.set_minor_locator(minorLocator)
ax1.yaxis.set_minor_locator(minorLocator)

ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')


fig.tight_layout()
# plt.subplots_adjust(wspace=0.25)
# plt.savefig("../plots/decimation-offset-dist.eps")
```

```python
# Normalized difference in RA and decl for all sources.
from scipy import stats
bins = np.linspace(-5, 5, 50)
weights1 = np.ones_like(nor_dRAc) * 100. / nor_dRAc.size

# Plot for the normalized difference
fig, (ax0, ax1) = plt.subplots(figsize=(8, 4), ncols=2)

ax0.hist(nor_dRAc, bins, weights=weights1)
ax0.set_xlabel("$X_{\\Delta\\alpha\cos\delta}$")

ax1.hist(nor_dDec, bins, weights=weights1)

ax0.set_ylabel("% in bin")
ax1.set_ylabel("% in bin")

ax0.set_xlabel("$X_{\\alpha*}$")
ax1.set_xlabel("$X_{\\delta}$")

ax0.set_ylabel("% in bin")
ax1.set_ylabel("% in bin")

# add a '(standard) Gaussian distribution' line
Gaussian_dist = stats.norm.pdf(bins, 0, 1) * 10. / 50
ax0.plot(bins, Gaussian_dist * 100, "r--", linewidth=1)
ax1.plot(bins, Gaussian_dist * 100, "r--", linewidth=1)

ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')

ax0.grid()
ax1.grid()

fig.tight_layout()
```

```python
# Normalized difference in RA and decl for VCS sources.
from scipy import stats
bins = np.linspace(-5, 5, 50)
weights2 = np.ones_like(nor_dRAc[sou_flg]) * 100. / nor_dRAc[sou_flg].size

# Plot for the normalized difference
fig, (ax0, ax1) = plt.subplots(figsize=(8, 4), ncols=2)

ax0.hist(nor_dRAc[sou_flg], bins, weights=weights2)
ax0.set_xlabel("$X_{\\Delta\\alpha\cos\delta}$")

ax1.hist(nor_dDec[sou_flg], bins, weights=weights2)

ax0.set_ylabel("% in bin")
ax1.set_ylabel("% in bin")

ax0.set_xlabel("$X_{\\alpha*}$")
ax1.set_xlabel("$X_{\\delta}$")

ax0.set_ylabel("% in bin")
ax1.set_ylabel("% in bin")

# add a '(standard) Gaussian distribution' line
Gaussian_dist = stats.norm.pdf(bins, 0, 1) * 10. / 50
ax0.plot(bins, Gaussian_dist * 100, "r--", linewidth=1)
ax1.plot(bins, Gaussian_dist * 100, "r--", linewidth=1)


ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')

ax0.grid()
ax1.grid()

fig.tight_layout()
plt.show()
```

```python
# Calculate the offset and normalized seperation
rho = sqrt(dRAc**2 + dDec**2)
nor_rho = sqrt(nor_dRAc**2 + nor_dDec**2)
rho_err = sqrt(dRAc_err**2 + dDec_err**2)

fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

bins = np.arange(0., 10., 0.5)

ax0.hist(rho, bins, histtype='bar', rwidth=0.8, label="All")
ax0.hist(rho[sou_flg], bins, histtype='bar', rwidth=0.8, label="VCS")
ax0.set_xlabel('$\\rho$ [mas]')
ax0.set_ylabel("Number of Sources")
ax0.legend()

bins = np.arange(0., 6., 0.25)

ax1.hist(X, bins, histtype='bar', rwidth=0.8, label="All")
ax1.hist(X[sou_flg], bins,
         histtype='bar', rwidth=0.8, label="VCS")
ax1.set_xlabel('$X$')
ax1.set_ylabel("Number of Sources")

from matplotlib.ticker import MultipleLocator, FormatStrFormatter
minorLocator = MultipleLocator(0.5)
ax0.xaxis.set_minor_locator(minorLocator)
ax1.xaxis.set_minor_locator(minorLocator)

minorLocator = MultipleLocator(50)
ax0.yaxis.set_minor_locator(minorLocator)
ax1.yaxis.set_minor_locator(minorLocator)

ax0.xaxis.set_ticks_position('both')
ax0.yaxis.set_ticks_position('both')
ax1.xaxis.set_ticks_position('both')
ax1.yaxis.set_ticks_position('both')

fig.tight_layout()
plt.subplots_adjust(wspace=0.2)
```
